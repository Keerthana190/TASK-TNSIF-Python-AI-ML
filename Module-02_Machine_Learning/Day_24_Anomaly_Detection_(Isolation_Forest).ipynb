# What is Anomaly Detection?

Anomaly Detection is an Unsupervised Machine Learning technique used to identify unusual or abnormal observations in a dataset.

These unusual observations are called anomalies or outliers.

Unlike normal data points, anomalies behave differently from the majority of the data.

## Real-Life Applications

- Credit Card Fraud Detection
- Network Intrusion Detection
- Medical Diagnosis
- Manufacturing Quality Control
- Sensor Failure Detection

# Isolation Forest

## Definition

Isolation Forest is an Unsupervised Machine Learning algorithm used to detect anomalies.

Instead of comparing data points with each other, it isolates observations using random decision trees.

Anomalies are isolated faster because they are rare and different from normal observations.



````python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.ensemble import IsolationForest
````

````python
# Sample Dataset

df = pd.DataFrame({

    "Transaction_Amount":[
        500,
        700,
        650,
        800,
        900,
        1000,
        1200,
        1500,
        50000,
        65000
    ]

})

df
````

````python
model = IsolationForest(

    contamination=0.2,

    random_state=42

)

df["Anomaly"] = model.fit_predict(df[["Transaction_Amount"]])

df
````

````python
print(df)
````

````python
normal = df[df["Anomaly"] == 1]

anomaly = df[df["Anomaly"] == -1]
````

````python
plt.figure(figsize=(7,4))

plt.scatter(

normal.index,

normal["Transaction_Amount"],

label="Normal"

)

plt.scatter(

anomaly.index,

anomaly["Transaction_Amount"],

label="Anomaly"

)

plt.xlabel("Transaction Number")

plt.ylabel("Transaction Amount")

plt.title("Isolation Forest - Anomaly Detection")

plt.legend()

plt.grid()

plt.show()
````

# Understanding the Output

The algorithm labels each observation as:

1 → Normal Data

-1 → Anomaly (Outlier)

Transactions with unusually high amounts are detected as anomalies.

# Advantages

- Detects anomalies automatically
- No labeled data required
- Fast and efficient
- Works well for large datasets
- Handles high-dimensional data

# Disadvantages

- Sensitive to contamination parameter
- Less effective when anomalies are very similar to normal data
- May require parameter tuning

|------------|------------------|

| Simple to implement | More powerful for complex datasets |
