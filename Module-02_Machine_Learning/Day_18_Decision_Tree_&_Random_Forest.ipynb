# Tree-Based Algorithms

Tree-based algorithms make decisions by splitting the dataset into smaller subsets based on feature values.

They are widely used because they are easy to understand and interpret.

## Types

- Decision Tree
- Random Forest

# Decision Tree

## Definition

Decision Tree is a supervised machine learning algorithm used for both classification and regression problems.

It makes predictions by creating a tree-like structure of decision rules.

## Real-Life Applications

- Loan Approval
- Disease Diagnosis
- Customer Segmentation
- Fraud Detection

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
# Sample Dataset

df = pd.DataFrame({

    "Age":[18,22,25,30,35,40,45,50],

    "Monthly_Income":[15000,20000,25000,35000,45000,55000,65000,75000],

    "Join_Gym":["No","No","Yes","Yes","Yes","No","No","No"]

})

df
```

```python
X = df[["Age","Monthly_Income"]]

y = df["Join_Gym"]
```

```python
model = DecisionTreeClassifier(random_state=42)

model.fit(X,y)
```

```python
prediction = model.predict([[28,30000]])

print("Prediction :",prediction[0])
```

```python
plt.figure(figsize=(10,6))

plot_tree(model,

feature_names=["Age","Monthly Income"],

class_names=["No","Yes"],

filled=True)

plt.title("Decision Tree")

plt.show()
```

## Advantages

- Easy to understand
- Easy to visualize
- Works with numerical and categorical data
- No feature scaling required

# Random Forest

## Definition

Random Forest is an ensemble learning algorithm that combines multiple Decision Trees and makes predictions based on majority voting.

It improves accuracy and reduces overfitting.

## Real-Life Applications

- Credit Card Fraud Detection
- Disease Prediction
- Customer Churn Prediction
- Recommendation Systems

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestClassifier
```

```python
# Sample Dataset

df = pd.DataFrame({

    "Age":[20,25,30,35,40,45,50,55],

    "Income":[25000,30000,40000,50000,60000,70000,80000,90000],

    "Purchased":["No","No","Yes","Yes","Yes","No","No","Yes"]

})

df
```

```python
X = df[["Age","Income"]]

y = df["Purchased"]
```

```python
model = RandomForestClassifier(n_estimators=100,random_state=42)

model.fit(X,y)
```

```python
prediction = model.predict([[32,45000]])

print("Prediction :",prediction[0])
```

```python
importance = pd.Series(

model.feature_importances_,

index=X.columns

)

importance.plot(kind="bar")

plt.title("Feature Importance")

plt.ylabel("Importance")

plt.show()
```

## Advantages

- Higher accuracy
- Reduces overfitting
- Handles large datasets
- Robust to noise

# Decision Tree vs Random Forest

| Decision Tree | Random Forest |
|--------------|---------------|
| Single Tree | Multiple Trees |
| Faster | Slightly Slower |
| Can Overfit | Less Overfitting |
| Easier to Interpret | More Accurate |
| Lower Accuracy | Higher Accuracy |
