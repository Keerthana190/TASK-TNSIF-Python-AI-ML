# What is Q-Learning?

Q-Learning is a Reinforcement Learning algorithm that helps an agent learn the best action to take in each state.

Unlike MDP, the agent does not need to know the environment beforehand.

Instead, it learns by repeatedly interacting with the environment and updating its knowledge after every action.

The objective is to maximize the total reward.

# Why Do We Need Q-Learning?

Imagine a robot entering a maze for the first time.

The robot does not know:

- Which path is correct
- Which path is shortest
- Where obstacles are located

It keeps trying different paths.

Whenever it reaches the goal, it receives a reward.

After many attempts, it learns the best route.

This learning process is called Q-Learning.

# Real-Life Example

Imagine you are visiting a new shopping mall.

On your first visit,

you don't know where the food court is.

You try different routes.

Some routes take longer.

Some routes are blocked.

Eventually, you remember the shortest route.

Next time, you directly choose the best route.

This is exactly how Q-Learning works.


# Important Terms

## State (S)

Current position of the agent.

Example:

Room A

Room B

Room C

---

## Action (A)

Possible movement.

Move Left

Move Right

Move Up

Move Down

---

## Reward (R)

Feedback after taking an action.

Positive Reward → Correct Action

Negative Reward → Wrong Action

---

## Q-Value

Q-Value tells how good an action is in a particular state.

Higher Q-Value means a better action.

# What is a Q-Table?

A Q-Table stores the Q-Values for every possible State and Action.

Example:

| State | Left | Right |
|--------|------|-------|
| A | 0 | 5 |
| B | 3 | 7 |
| C | 2 | 10 |

The agent always tries to choose the action with the highest Q-Value.

# Bellman Equation

Q-Learning updates its knowledge using the Bellman Equation.

Q(New) = Q(Old) + α [Reward + γ(Max Future Q) − Q(Old)]

Where:

α (Alpha) = Learning Rate

γ (Gamma) = Discount Factor

Reward = Immediate reward received

Max Future Q = Best future reward

# Learning Rate (α)

Learning Rate decides how quickly the agent learns.

Small α

Learning is slow but stable.

Large α

Learning is faster but may become unstable.

# Discount Factor (γ)

Discount Factor decides how important future rewards are.

γ close to 1

Future rewards are important.

γ close to 0

Only immediate rewards matter.

# Exploration vs Exploitation

## Exploration

The agent tries new actions to discover better solutions.

Example:

Trying a new road to school.

## Exploitation

The agent uses what it has already learned.

Example:

Taking the shortest known route every day.

A good Reinforcement Learning agent balances both.

```python
import pandas as pd

# Simple Q-Table

q_table = pd.DataFrame({

    "Left":[0,2,1],

    "Right":[5,7,10]

},

index=["State A","State B","State C"])

q_table
```

```python
best_action = q_table.idxmax(axis=1)

print("Best Action for Each State:\n")

print(best_action)
```

```python
import matplotlib.pyplot as plt

q_table.plot(kind="bar")

plt.title("Q-Values for Different Actions")

plt.xlabel("States")

plt.ylabel("Q-Value")

plt.grid()

plt.show()
```

# Advantages

✔ Learns without labeled data

✔ Finds the best action automatically

✔ Suitable for sequential decision making

✔ Easy to understand and implement

✔ Works well in unknown environments

# Disadvantages

✖ Learning can be slow

✖ Large state spaces require large Q-Tables

✖ Exploration takes time

✖ Not suitable for very complex environments

# MDP vs Q-Learning

| MDP | Q-Learning |
|------|------------|
| Mathematical framework | Learning algorithm |
| Requires environment model | Learns from experience |
| Uses transition probabilities | Does not require transition probabilities |
| Describes the problem | Solves the problem |
