# Support Vector Machine (SVM)

## Definition

Support Vector Machine (SVM) is a supervised machine learning algorithm used mainly for classification problems.

It finds the best decision boundary (called a hyperplane) that separates different classes.

## Real-Life Applications

- Face Recognition
- Text Classification
- Spam Detection
- Image Classification

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.svm import SVC
```

```python
# Sample Dataset

df = pd.DataFrame({

    "Age":[20,22,25,28,30,35,40,45],

    "Income":[20000,25000,30000,35000,40000,50000,60000,70000],

    "Purchased":["No","No","No","Yes","Yes","Yes","No","No"]

})

df
```

```python
X = df[["Age","Income"]]

y = df["Purchased"]
```

```python
model = SVC(kernel="linear")

model.fit(X,y)
```

```python
prediction = model.predict([[29,38000]])

print("Prediction :",prediction[0])
```

```python
plt.scatter(df["Age"],df["Income"])

plt.xlabel("Age")

plt.ylabel("Income")

plt.title("SVM Dataset")

plt.grid()

plt.show()
```

## Advantages

- Works well with high-dimensional data
- Effective for small and medium datasets
- Good accuracy
- Can solve linear and non-linear problems

# Naive Bayes

## Definition

Naive Bayes is a supervised machine learning algorithm based on Bayes' Theorem.

It assumes that all features are independent of each other.

## Real-Life Applications

- Spam Email Detection
- Sentiment Analysis
- News Classification
- Document Categorization

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.naive_bayes import GaussianNB
```

```python
# Sample Dataset

df = pd.DataFrame({

    "Age":[22,25,28,30,35,40,45,50],

    "Salary":[25000,30000,35000,45000,55000,65000,75000,85000],

    "Purchased":["No","No","Yes","Yes","Yes","No","No","Yes"]

})

df
```

```python
X = df[["Age","Salary"]]

y = df["Purchased"]
```

```python
model = GaussianNB()

model.fit(X,y)
```

```python
prediction = model.predict([[32,50000]])

print("Prediction :",prediction[0])
```

```python
plt.scatter(df["Age"],df["Salary"])

plt.xlabel("Age")

plt.ylabel("Salary")

plt.title("Naive Bayes Dataset")

plt.grid()

plt.show()
```

## Advantages

- Very fast
- Easy to implement
- Works well with small datasets
- Excellent for text classification

# SVM vs Naive Bayes

| Support Vector Machine | Naive Bayes |
|-------------------------|-------------|
| Finds the best hyperplane | Uses Bayes' Theorem |
| Higher accuracy for complex data | Faster training |
| Works well in high-dimensional data | Works well for text data |
| Training is slower | Training is very fast |
| More computationally expensive | Less computationally expensive |
