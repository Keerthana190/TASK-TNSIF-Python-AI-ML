# What is Markov Decision Process (MDP)?

A Markov Decision Process (MDP) is a mathematical framework used in Reinforcement Learning to help an agent make the best decisions in an environment.

It describes how an agent moves from one state to another by taking actions and receiving rewards.

The goal of the agent is to maximize the total reward over time.

# Why Do We Need MDP?

In Reinforcement Learning, the agent must make many decisions.

Instead of choosing actions randomly, the agent needs a structured way to decide:

- Where am I now?
- What actions can I take?
- What reward will I receive?
- Which action is the best?

MDP provides this decision-making framework.

# Real-Life Example

Imagine you are using Google Maps.

Current Location → State

Choose Left or Right → Action

Reach Destination Faster → Reward

Traffic Condition → Environment

Google Maps continuously decides the best route to maximize your benefit (less time).

This is similar to how an MDP works.

# Components of MDP

## 1. State (S)

Represents the current situation of the agent.

Example:
Room A
Room B
Room C

---

## 2. Action (A)

The possible decisions taken by the agent.

Example:

Move Left

Move Right

Move Forward

---

## 3. Reward (R)

Feedback received after taking an action.

Positive Reward → Good Action

Negative Reward → Bad Action

---

## 4. Transition Probability (P)

The probability of moving from one state to another after taking an action.

Example:

There is an 80% chance of reaching Room B.

There is a 20% chance of reaching Room C.

---

## 5. Policy (π)

A strategy that tells the agent which action to choose in every state.

# Workflow of MDP

Current State

↓

Choose an Action

↓

Move to Next State

↓

Receive Reward

↓

Update Policy

↓

Repeat Until Goal is Reached

# MDP Formula

An MDP is represented as:

MDP = (S, A, P, R, γ)

Where,

S → States

A → Actions

P → Transition Probability

R → Reward

γ → Discount Factor

The Discount Factor (γ) decides how much importance is given to future rewards.

```python
import pandas as pd

df = pd.DataFrame({

    "Current_State":["Home","Road","School"],

    "Action":["Drive","Continue","Stop"],

    "Next_State":["Road","School","School"],

    "Reward":[2,5,10]

})

df
```

```python
for i in range(len(df)):

    print("Current State :",df.loc[i,"Current_State"])

    print("Action :",df.loc[i,"Action"])

    print("Next State :",df.loc[i,"Next_State"])

    print("Reward :",df.loc[i,"Reward"])

    print("----------------------")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

plt.plot(

df["Current_State"],

df["Reward"],

marker="o"

)

plt.title("Reward in Each State")

plt.xlabel("State")

plt.ylabel("Reward")

plt.grid()

plt.show()
```
