# What is Unsupervised Learning?

## Definition

Unsupervised Learning is a type of Machine Learning where the model learns from **unlabeled data**.

Unlike Supervised Learning, there is **no target/output variable**. The algorithm automatically discovers hidden patterns, similarities, and relationships within the data.

### Formula

Input Data (X) → Unsupervised Algorithm → Hidden Patterns / Groups

## Real-Life Applications

- Customer Segmentation
- Product Recommendation
- Fraud Detection
- Market Basket Analysis
- Face Recognition
- Image Segmentation

# Characteristics of Unsupervised Learning

- Uses unlabeled data
- No target/output variable
- Finds hidden patterns
- Groups similar data points
- Used for clustering and dimensionality reduction
- No manual labeling required

# K-Means Clustering

## Definition

K-Means is one of the most popular clustering algorithms.

It divides the dataset into K clusters based on similarity.

## Steps

1. Choose K
2. Select Initial Centroids
3. Assign Points to the Nearest Centroid
4. Update Centroids
5. Repeat Until Clusters Stop Changing


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = {
    "Experience":[1,2,3,5,6,8,10,12,15,18],
    "Salary":[25000,28000,32000,45000,50000,65000,80000,90000,110000,130000]
}

df = pd.DataFrame(data)

print(df)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)

df["Cluster"] = kmeans.fit_predict(df)

print(df)
```

```python
plt.figure(figsize=(6,4))

plt.scatter(df["Experience"],
            df["Salary"],
            c=df["Cluster"],
            s=100)

plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1],
            color="red",
            marker="X",
            s=250)

plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Employee Salary Clustering")
plt.grid(True)

plt.show()
```

# Hierarchical Clustering

## Definition

Hierarchical Clustering groups similar data points into a hierarchy of clusters.

It creates a tree-like structure called a Dendrogram.

Unlike K-Means, it does not require specifying the number of clusters beforehand.

## Applications

- Gene Analysis
- Customer Segmentation
- Document Clustering

# DBSCAN

## Definition

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) groups nearby data points based on density.

It can detect noise (outliers) and does not require specifying the number of clusters.

## Parameters

- eps
- min_samples

## Applications

- Fraud Detection
- GPS Data Analysis
- Anomaly Detection

```python
from sklearn.cluster import DBSCAN
```

```python
data = {
    "Amount":[100,120,150,180,200,220,250,280,300,5000],
    "Transactions":[5,6,5,7,8,6,7,8,9,1]
}

df = pd.DataFrame(data)

print(df)
```

```python
dbscan = DBSCAN(eps=100, min_samples=2)

df["Cluster"] = dbscan.fit_predict(df)

print(df)
```

```python
plt.figure(figsize=(6,4))

plt.scatter(df["Amount"],
            df["Transactions"],
            c=df["Cluster"],
            s=100)

plt.xlabel("Transaction Amount")
plt.ylabel("Number of Transactions")
plt.title("Fraud Detection using DBSCAN")
plt.grid(True)

plt.show()
```

# Difference Between Clustering Algorithms

| Feature | K-Means | Hierarchical | DBSCAN |
|----------|----------|--------------|---------|
| Requires K | Yes | No | No |
| Method | Centroid-Based | Tree-Based | Density-Based |
| Detects Noise | No | Limited | Yes |
| Speed | Fast | Moderate | Moderate |
| Best For | Large Datasets | Small Datasets | Outlier Detection |
