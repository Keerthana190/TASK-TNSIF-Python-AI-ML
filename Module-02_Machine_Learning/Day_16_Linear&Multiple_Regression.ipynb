## What is Regression?

Regression is a Supervised Machine Learning algorithm used to predict continuous numerical values.

It learns the relationship between input features (X) and the output variable (Y) and predicts numerical values for new data.

### Real-Life Applications

- House Price Prediction
- Salary Prediction
- Sales Forecasting
- Temperature Prediction
- Stock Price Prediction

# Simple Linear Regression

### Definition

Simple Linear Regression predicts a continuous value using one independent variable.

### Formula

y = mx + c

Where

- x = Independent Variable
- y = Dependent Variable
- m = Slope
- c = Intercept

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Sample Dataset

df = pd.DataFrame({

    "House_Size":[500,700,900,1100,1300,1500,1700,2100,2300],

    "House_Price":[25,35,45,55,65,75,85,105,115]

})

df
```

```python
# Scatter Plot

plt.scatter(df["House_Size"],df["House_Price"])

plt.xlabel("House Size")

plt.ylabel("House Price")

plt.title("House Size vs House Price")

plt.grid()

plt.show()
```

```python
# Independent Variable

X=df[["House_Size"]]

# Dependent Variable

y=df["House_Price"]
```

```python
# Create Model

model=LinearRegression()

# Train Model

model.fit(X,y)
```

```python
# Prediction

prediction=model.predict([[1250]])

print("Predicted House Price :",prediction[0])

print("Slope :",model.coef_[0])

print("Intercept :",model.intercept_)
```

```python
# Regression Line

plt.scatter(df["House_Size"],df["House_Price"],label="Actual Data")

plt.plot(df["House_Size"],model.predict(X),color="red",label="Regression Line")

plt.xlabel("House Size")

plt.ylabel("House Price")

plt.title("Simple Linear Regression")

plt.legend()

plt.show()
```

### Advantages

- Easy to implement
- Easy to understand
- Fast training
- Suitable for linear relationships

# Multiple Linear Regression

### Definition

Multiple Linear Regression predicts a continuous value using two or more independent variables.

### Formula

y = b₀ + b₁x₁ + b₂x₂ + ... + bₙxₙ

### Example

Rainfall + Fertilizer → Crop Yield

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Sample Dataset

df=pd.DataFrame({

    "Rainfall":[500,550,600,650,700,750,800],

    "Fertilizer":[40,45,50,55,60,65,70],

    "Crop_Yield":[2.5,2.8,3.2,3.6,4.0,4.4,4.8]

})

df
```

```python
# Independent Variables

X=df[["Rainfall","Fertilizer"]]

# Dependent Variable

y=df["Crop_Yield"]
```

```python
# Create Model

model=LinearRegression()

# Train Model

model.fit(X,y)
```

```python
# Prediction

prediction=model.predict([[720,62]])

print("Predicted Crop Yield :",prediction[0])

print()

print("Rainfall Coefficient :",model.coef_[0])

print("Fertilizer Coefficient :",model.coef_[1])

print("Intercept :",model.intercept_)
```

```python
# Rainfall vs Crop Yield

plt.scatter(df["Rainfall"],df["Crop_Yield"])

plt.xlabel("Rainfall")

plt.ylabel("Crop Yield")

plt.title("Rainfall vs Crop Yield")

plt.grid()

plt.show()
```

```python
# Fertilizer vs Crop Yield

plt.scatter(df["Fertilizer"],df["Crop_Yield"])

plt.xlabel("Fertilizer")

plt.ylabel("Crop Yield")

plt.title("Fertilizer vs Crop Yield")

plt.grid()

plt.show()
```

### Advantages

- Uses multiple features
- Better prediction accuracy
- Suitable for real-world datasets
- Captures relationships among multiple variables

# Difference Between Simple and Multiple Linear Regression

| Simple Linear Regression | Multiple Linear Regression |
|---------------------------|----------------------------|
| Uses one independent variable | Uses two or more independent variables |
| Easy to visualize | Harder to visualize |
| Simpler model | More complex model |
| Lower complexity | Higher complexity |

# Polynomial Regression

## Definition

Polynomial Regression is an extension of Linear Regression that models the relationship between variables as a polynomial equation instead of a straight line.

It is useful when the data follows a curved pattern rather than a linear pattern.

## Formula

y = b₀ + b₁x + b₂x² + ... + bₙxⁿ

## Real-Life Applications

- Population Growth
- Disease Spread
- Sales Forecasting
- Temperature Prediction

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Sample Dataset

df = pd.DataFrame({

    "Years":[1,2,3,4,5,6,7,8],

    "Salary":[25,28,35,50,70,95,130,170]

})

df
```

```python
plt.scatter(df["Years"],df["Salary"])

plt.title("Years vs Salary")

plt.xlabel("Years")

plt.ylabel("Salary")

plt.grid()

plt.show()
```

```python
X=df[["Years"]]

y=df["Salary"]
```

```python
poly=PolynomialFeatures(degree=2)

X_poly=poly.fit_transform(X)
```

```python
model=LinearRegression()

model.fit(X_poly,y)
```

```python
prediction=model.predict(poly.transform([[9]]))

print("Predicted Salary :",prediction[0])
```

```python
plt.scatter(df["Years"],df["Salary"])

plt.plot(X,
         model.predict(X_poly),
         color="red")

plt.title("Polynomial Regression")

plt.xlabel("Years")

plt.ylabel("Salary")

plt.grid()

plt.show()
```

## Advantages

- Captures curved relationships
- Better accuracy than linear regression for non-linear data
- Easy to implement

| Feature          | Simple Linear Regression | Multiple Linear Regression        | Polynomial Regression        |
| ---------------- | ------------------------ | --------------------------------- | ---------------------------- |
| Number of Inputs | One                      | Two or More                       | One or More                  |
| Relationship     | Linear                   | Linear                            | Non-Linear                   |
| Equation         | Y = mX + c               | Y = b₀+b₁X₁+b₂X₂+...              | Y = a+bX+cX²+...             |
| Best For         | Simple relationships     | Multiple factors affecting output | Curved relationships         |
| Graph            | Straight Line            | Multi-dimensional Plane           | Curved Line                  |
| Example          | House Size → Price       | Size + Bedrooms → Price           | Temperature → Sales (Curved) |
