# What is Deep Q-Network (DQN)?

Deep Q-Network (DQN) is an advanced Reinforcement Learning algorithm.

It combines:

- Reinforcement Learning
- Deep Neural Networks

Instead of storing values in a Q-Table, DQN uses a Neural Network to predict Q-values.

This allows DQN to solve much larger and more complex problems than traditional Q-Learning.

# Why Do We Need DQN?

Q-Learning stores values in a Q-Table.

If there are millions of states, the Q-Table becomes too large.

Example:

Imagine a self-driving car.

Possible situations include:

- Traffic
- Signals
- Speed
- Weather
- Pedestrians

Creating a Q-Table for all these situations is impossible.

Instead, DQN uses a Neural Network to estimate the best action.

# Real-Life Example

Imagine learning to play a video game.

At first, you make random moves.

After playing many times, your brain remembers which actions give higher scores.

Instead of memorizing every situation, your brain recognizes patterns.

Similarly, DQN uses a Neural Network to recognize patterns and predict the best action.

# How DQN Works

1. Observe the current state.

2. Feed the state into a Neural Network.

3. Predict Q-values for all possible actions.

4. Choose the best action.

5. Receive a reward.

6. Update the Neural Network.

7. Repeat until the best policy is learned.

# Important Concepts

## 1. Neural Network

Instead of using a Q-Table, DQN uses a Neural Network to estimate Q-values.

---

## 2. Experience Replay

The agent stores previous experiences in memory.

During training, it randomly selects past experiences.

This improves learning and prevents the model from depending only on recent experiences.

---

## 3. Target Network

A separate neural network is used to generate stable target Q-values.

This helps make training more stable.

# DQN Workflow

Current State

↓

Neural Network predicts Q-values

↓

Choose Best Action

↓

Perform Action

↓

Receive Reward

↓

Store Experience

↓

Train Neural Network

↓

Repeat

```python
import pandas as pd

# Sample Data

df = pd.DataFrame({

    "State":["S1","S2","S3","S4"],

    "Action":["Left","Right","Forward","Stop"],

    "Predicted_Q_Value":[2.3,5.8,8.5,10.0]

})

df
```

```python
best = df.loc[df["Predicted_Q_Value"].idxmax()]

print("Best State :",best["State"])

print("Best Action :",best["Action"])

print("Predicted Q-Value :",best["Predicted_Q_Value"])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

plt.bar(df["State"],df["Predicted_Q_Value"])

plt.title("Predicted Q-Values using DQN")

plt.xlabel("State")

plt.ylabel("Q-Value")

plt.grid()

plt.show()
```

# Q-Learning vs Deep Q-Network (DQN)

| Q-Learning | DQN |
|------------|-----|
| Uses Q-Table | Uses Neural Network |
| Suitable for small problems | Suitable for large problems |
| Simple implementation | More complex implementation |
| Limited state space | Handles high-dimensional state spaces |
| Faster for small tasks | Better for complex tasks |
