# What is Association Rule Learning?

## Definition

Association Rule Learning is an Unsupervised Machine Learning technique used to discover interesting relationships or associations between items in a dataset.

It answers questions like:

- Which items are frequently purchased together?
- What products are likely to be bought if another product is purchased?

## Example

- Bread → Butter
- Laptop → Mouse
- Mobile → Earphones

## Real-Life Applications

- Market Basket Analysis
- Product Recommendation
- Cross Selling
- Retail Stores

# Important Terms

## 1. Support

Support tells how frequently an item or itemset appears in the dataset.

Support(A) = Number of transactions containing A / Total number of transactions

Example:

Milk appears in 4 out of 5 transactions.

Support = 4/5 = 0.8

---

## 2. Confidence

Confidence tells how often item B is purchased when item A is purchased.

Confidence(A → B) = Support(A ∩ B) / Support(A)

Example:

Milk appears = 4

Milk & Bread together = 3

Confidence = 3/4 = 0.75

---

## 3. Lift

Lift measures the strength of the association between two items.

Lift(A → B) = Confidence(A → B) / Support(B)

Higher Lift indicates a stronger relationship.

# Apriori Algorithm

## Definition

Apriori is one of the most popular Association Rule Learning algorithms used to discover frequent itemsets and generate association rules.

### Principle

"If an itemset is frequent, then all of its subsets must also be frequent."

## Applications

- Market Basket Analysis
- Product Recommendation
- Sales Analysis


```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
transactions = [
    ['Milk','Bread','Butter'],
    ['Milk','Bread'],
    ['Milk','Butter'],
    ['Bread','Butter'],
    ['Milk','Bread','Butter'],
    ['Milk','Eggs'],
    ['Bread','Eggs'],
    ['Milk','Bread','Eggs']
]
```

```python
te = TransactionEncoder()

te_array = te.fit(transactions).transform(transactions)

df = pd.DataFrame(te_array, columns=te.columns_)

print(df)
```

```python
frequent_itemsets = apriori(
    df,
    min_support=0.3,
    use_colnames=True
)

print(frequent_itemsets)
```

```python
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.6
)

print(rules[[
    "antecedents",
    "consequents",
    "support",
    "confidence",
    "lift"
]])
```

# Advantages

# Disadvantages

- Generates many rules for large datasets
- Can be computationally expensive
- Requires setting minimum support and confidence values carefully
