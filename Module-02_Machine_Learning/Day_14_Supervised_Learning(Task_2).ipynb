```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Dataset
df = pd.DataFrame({
    "Temperature": [20, 22, 24, 26, 28, 30, 32],
    "IceCreamSales": [100, 120, 140, 160, 180, 200, 220]
})

# Independent Variable
X = df[["Temperature"]]

# Dependent Variable
y = df["IceCreamSales"]

# Create Model
model = LinearRegression()

# Train Model
model.fit(X, y)

# Prediction
prediction = model.predict([[35]])

print("Predicted Ice Cream Sales at 35°C:", prediction[0])

# Model Parameters
print("\nSlope:", model.coef_[0])
print("Intercept:", model.intercept_)

# Accuracy
score = model.score(X, y)
print("R² Score:", score)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

# Actual Data
plt.scatter(
    df["Temperature"],
    df["IceCreamSales"],
    label="Actual Data"
)

# Regression Line
plt.plot(
    df["Temperature"],
    model.predict(X),
    label="Regression Line"
)

plt.title("Temperature vs Ice Cream Sales")
plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Cream Sales")
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.scatter(y, model.predict(X))

plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.title("Actual vs Predicted Sales")

plt.grid(True)

plt.show()
```

```python
# Import Libraries
import pandas as pd
from sklearn.linear_model import LinearRegression

# Create Dataset
df = pd.DataFrame({
    "Study_Hours": [2, 3, 4, 5, 6, 7, 8],
    "Attendance": [60, 65, 70, 75, 80, 85, 90],
    "Marks": [45, 50, 58, 65, 72, 80, 88]
})

# Display Dataset
print("Dataset")
print(df)

# Independent Variables (X)
X = df[["Study_Hours", "Attendance"]]

# Dependent Variable (Y)
y = df["Marks"]

# Create Model
model = LinearRegression()

# Train Model
model.fit(X, y)

# Prediction
prediction = model.predict([[6, 82]])

print("\nPredicted Marks:", round(prediction[0], 2))

# Model Coefficients
print("\nCoefficient for Study Hours:", model.coef_[0])
print("Coefficient for Attendance:", model.coef_[1])
print("Intercept:", model.intercept_)

# R² Score
print("\nModel Accuracy (R² Score):", model.score(X, y))
```

```python
import matplotlib.pyplot as plt

# Study Hours vs Marks
plt.figure(figsize=(6,4))

plt.scatter(df["Study_Hours"], df["Marks"])

plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.grid(True)

plt.show()


# Attendance vs Marks
plt.figure(figsize=(6,4))

plt.scatter(df["Attendance"], df["Marks"])

plt.title("Attendance vs Marks")
plt.xlabel("Attendance (%)")
plt.ylabel("Marks")

plt.grid(True)

plt.show()
```

```python
# Import Libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Create Dataset
df = pd.DataFrame({
    "Study_Hours": [1, 2, 3, 4, 5, 6, 7, 8],
    "Marks": [15, 25, 40, 60, 78, 90, 96, 99]
})

# Display Dataset
print("Dataset")
print(df)

# Independent Variable
X = df[["Study_Hours"]]

# Dependent Variable
y = df["Marks"]

# Convert Linear Features to Polynomial Features
poly = PolynomialFeatures(degree=2)

X_poly = poly.fit_transform(X)

# Create Model
model = LinearRegression()

# Train Model
model.fit(X_poly, y)

# Prediction
prediction = model.predict(poly.transform([[5.5]]))

print("\nPredicted Marks for 5.5 Study Hours:", round(prediction[0],2))

# Model Coefficients
print("\nCoefficients:", model.coef_)
print("Intercept:", model.intercept_)

# Model Accuracy
print("\nR² Score:", model.score(X_poly, y))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate Smooth Curve
X_grid = np.arange(min(X.values), max(X.values), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)

# Plot
plt.figure(figsize=(6,4))

# Actual Data
plt.scatter(df["Study_Hours"], df["Marks"], label="Actual Data")

# Polynomial Curve
plt.plot(
    X_grid,
    model.predict(poly.transform(X_grid)),
    label="Polynomial Regression Curve"
)

plt.title("Polynomial Regression")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor

# Dataset
df = pd.DataFrame({
    "Experience": [1, 2, 3, 4, 5, 6, 7, 8],
    "Salary": [25000, 30000, 40000, 50000, 65000, 80000, 95000, 110000]
})

# Independent Variable
X = df[["Experience"]]

# Dependent Variable
y = df["Salary"]

# Create Decision Tree Regression Model
model = DecisionTreeRegressor(random_state=42)

# Train Model
model.fit(X, y)

# Prediction
prediction = model.predict([[5.5]])

print("Predicted Salary:", prediction[0])

# Model Accuracy
print("R² Score:", model.score(X, y))
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# Dataset
df = pd.DataFrame({
    "Age": [22, 25, 28, 30, 35, 40, 45, 50],
    "Monthly_Income": [20000, 25000, 30000, 40000, 50000, 60000, 70000, 80000],
    "Credit_Score": [600, 620, 650, 700, 720, 750, 780, 800],
    "Loan_Approved": ["No", "No", "Yes", "Yes", "Yes", "Yes", "No", "No"]
})

# Independent Variables
X = df[["Age", "Monthly_Income", "Credit_Score"]]

# Dependent Variable
y = df["Loan_Approved"]

# Create Random Forest Model
model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

# Train Model
model.fit(X, y)

# Prediction
prediction = model.predict([[32, 45000, 710]])

print("Loan Approval Prediction:", prediction[0])

# Feature Importance
importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": model.feature_importances_
})

print("\nFeature Importance:")
print(importance.sort_values(by="Importance", ascending=False))
```

```python
import matplotlib.pyplot as plt

importance = importance.sort_values(by="Importance")

plt.figure(figsize=(6,4))

plt.barh(
    importance["Feature"],
    importance["Importance"]
)

plt.title("Random Forest Feature Importance")
plt.xlabel("Importance Score")
plt.ylabel("Features")

plt.grid(axis="x")

plt.show()
```

```python
import pandas as pd
from sklearn.svm import SVC

# Dataset
df = pd.DataFrame({
    "Age": [22, 25, 28, 30, 35, 40, 45, 50],
    "Monthly_Income": [20000, 25000, 30000, 35000, 45000, 55000, 65000, 75000],
    "Purchased": ["No", "No", "No", "Yes", "Yes", "Yes", "Yes", "No"]
})

# Independent Variables
X = df[["Age", "Monthly_Income"]]

# Dependent Variable
y = df["Purchased"]

# Create SVM Model
model = SVC(kernel="linear")

# Train Model
model.fit(X, y)

# Prediction
prediction = model.predict([[32, 40000]])

print("Purchase Prediction:", prediction[0])

# Accuracy
accuracy = model.score(X, y)
print("Model Accuracy:", round(accuracy * 100, 2), "%")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Convert labels to numeric values
y_numeric = [0 if i == "No" else 1 for i in y]

# Create mesh grid
x_min, x_max = X["Age"].min()-2, X["Age"].max()+2
y_min, y_max = X["Monthly_Income"].min()-5000, X["Monthly_Income"].max()+5000

xx, yy = np.meshgrid(
    np.arange(x_min, x_max, 0.2),
    np.arange(y_min, y_max, 500)
)

# Predict over mesh grid
Z = model.predict(pd.DataFrame({
    "Age": xx.ravel(),
    "Monthly_Income": yy.ravel()
}))

Z = [0 if i == "No" else 1 for i in Z]
Z = np.array(Z).reshape(xx.shape)

# Plot decision boundary
plt.figure(figsize=(7,5))

plt.contourf(xx, yy, Z, alpha=0.3)

plt.scatter(
    X["Age"],
    X["Monthly_Income"],
    c=y_numeric,
    s=100
)

# New Customer
plt.scatter(
    32,
    40000,
    marker="*",
    s=250,
    label="New Customer"
)

plt.title("Support Vector Machine (SVM)")
plt.xlabel("Age")
plt.ylabel("Monthly Income")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

# Dataset
df = pd.DataFrame({
    "Age": [20, 22, 25, 28, 30, 35, 40, 45],
    "Monthly_Income": [18000, 22000, 25000, 30000, 35000, 50000, 65000, 80000],
    "Purchased": ["No", "No", "No", "Yes", "Yes", "Yes", "Yes", "No"]
})

# Independent Variables
X = df[["Age", "Monthly_Income"]]

# Dependent Variable
y = df["Purchased"]

# Create KNN Model
model = KNeighborsClassifier(n_neighbors=3)

# Train Model
model.fit(X, y)

# Prediction
prediction = model.predict([[29, 32000]])

print("Purchase Prediction:", prediction[0])

# Accuracy
accuracy = model.score(X, y)
print("Model Accuracy:", round(accuracy * 100, 2), "%")
```

```python
import matplotlib.pyplot as plt

# Convert labels into numeric values for plotting
colors = [0 if i == "No" else 1 for i in y]

plt.figure(figsize=(6,4))

plt.scatter(
    df["Age"],
    df["Monthly_Income"],
    c=colors,
    s=100
)

# New customer
plt.scatter(
    29,
    32000,
    marker="*",
    s=250,
    label="New Customer"
)

plt.xlabel("Age")
plt.ylabel("Monthly Income")
plt.title("KNN Classification")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

# Dataset
df = pd.DataFrame({
    "Study_Hours": [1, 2, 3, 4, 5, 6, 7, 8],
    "Pass": [0, 0, 0, 0, 1, 1, 1, 1]
})

# Independent Variable
X = df[["Study_Hours"]]

# Dependent Variable
y = df["Pass"]

# Create Logistic Regression Model
model = LogisticRegression()

# Train Model
model.fit(X, y)

# Prediction
prediction = model.predict([[5]])

print("Pass Prediction:", "Pass" if prediction[0] == 1 else "Fail")

# Probability
probability = model.predict_proba([[5]])

print("Probability of Fail:", round(probability[0][0], 2))
print("Probability of Pass:", round(probability[0][1], 2))

# Accuracy
accuracy = model.score(X, y)
print("Model Accuracy:", round(accuracy * 100, 2), "%")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate values for sigmoid curve
x_values = np.linspace(0, 9, 200).reshape(-1, 1)
y_prob = model.predict_proba(x_values)[:, 1]

plt.figure(figsize=(6,4))

# Original Data
plt.scatter(df["Study_Hours"], df["Pass"], label="Actual Data")

# Sigmoid Curve
plt.plot(x_values, y_prob, label="Sigmoid Curve")

plt.xlabel("Study Hours")
plt.ylabel("Probability of Passing")
plt.title("Logistic Regression")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Dataset
df = pd.DataFrame({
    "Age": [22, 25, 28, 30, 35, 40, 45, 50],
    "Monthly_Income": [20000, 25000, 30000, 35000, 45000, 55000, 65000, 75000],
    "Purchased": ["No", "No", "No", "Yes", "Yes", "Yes", "Yes", "No"]
})

# Independent Variables
X = df[["Age", "Monthly_Income"]]

# Dependent Variable
y = df["Purchased"]

# Create Naive Bayes Model
model = GaussianNB()

# Train Model
model.fit(X, y)

# Prediction
prediction = model.predict([[32, 40000]])

print("Purchase Prediction:", prediction[0])

# Predictions for all records
y_pred = model.predict(X)

# Accuracy
accuracy = accuracy_score(y, y_pred)
print("Model Accuracy:", round(accuracy * 100, 2), "%")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

# Display Confusion Matrix
ConfusionMatrixDisplay.from_estimator(
    model,
    X,
    y,
    cmap="Blues"
)

plt.title("Naive Bayes - Confusion Matrix")
plt.show()
```
