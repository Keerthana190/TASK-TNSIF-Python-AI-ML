# What is Classification?

Classification is a type of Supervised Machine Learning in which the model predicts categorical values instead of numerical values.

The model learns from labeled data and predicts the category of new data.

### Examples

- Pass / Fail
- Spam / Not Spam
- Disease / No Disease
- Yes / No
- Fraud / Not Fraud

### Real-Life Applications

- Email Spam Detection
- Loan Approval
- Medical Diagnosis
- Face Recognition
- Customer Churn Prediction

# Regression vs Classification

| Regression | Classification |
|------------|----------------|
| Predicts numerical values | Predicts categories |
| Output is continuous | Output is discrete |
| Example: House Price | Example: Spam Detection |
| Algorithms: Linear Regression | Algorithms: Logistic Regression, KNN, SVM |

# Logistic Regression

## Definition

Logistic Regression is a supervised learning algorithm used for binary classification problems.

Although its name contains "Regression", it is mainly used for Classification.

### Uses

- Disease Prediction
- Email Spam Detection
- Loan Approval
- Student Pass/Fail Prediction

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
# Sample Dataset

df = pd.DataFrame({

    "Study_Hours":[1,2,3,4,5,6,7,8],

    "Pass":[0,0,0,0,1,1,1,1]

})

df
```

```python
X = df[["Study_Hours"]]

y = df["Pass"]
```

```python
model = LogisticRegression()

model.fit(X,y)
```

```python
prediction = model.predict([[5]])

print("Prediction :",prediction[0])

probability = model.predict_proba([[5]])

print("Probability :",probability)
```

```python
plt.scatter(df["Study_Hours"],df["Pass"])

plt.xlabel("Study Hours")

plt.ylabel("Pass")

plt.title("Logistic Regression")

plt.grid()

plt.show()
```

## Advantages

- Easy to implement
- Fast training
- Works well for binary classification
- Provides probability values

# K-Nearest Neighbors (KNN)

## Definition

K-Nearest Neighbors (KNN) is a supervised machine learning algorithm used for classification and regression.

It classifies a new data point based on the majority class of its nearest neighbors.

### Real-Life Applications

- Recommendation Systems
- Image Recognition
- Medical Diagnosis
- Pattern Recognition

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Sample Dataset

df = pd.DataFrame({

    "Age":[18,20,22,25,30,35,40,45],

    "Purchased":[0,0,0,1,1,1,0,0]

})

df
```

```python
X = df[["Age"]]

y = df["Purchased"]
```

```python
model = KNeighborsClassifier(n_neighbors=3)

model.fit(X,y)
```

```python
prediction = model.predict([[28]])

print("Prediction :",prediction[0])
```

```python
plt.scatter(df["Age"],df["Purchased"])

plt.xlabel("Age")

plt.ylabel("Purchased")

plt.title("KNN Classification")

plt.grid()

plt.show()
```

## Advantages

- Simple and easy to understand
- No training phase
- Good for small datasets
- Works well with non-linear data

# Logistic Regression vs KNN

| Logistic Regression | KNN |
|---------------------|-----|
| Parametric Algorithm | Non-Parametric Algorithm |
| Uses Sigmoid Function | Uses Distance Measure |
| Fast Prediction | Slower Prediction |
| Works well on large datasets | Better for small datasets |
| Requires model training | Lazy Learning Algorithm |
