# What is Reinforcement Learning?

Reinforcement Learning (RL) is a type of Machine Learning in which an agent learns by interacting with its environment.

Unlike Supervised Learning, the agent is not given correct answers. Instead, it learns through trial and error.

Whenever the agent performs a good action, it receives a reward.

If it performs a wrong action, it receives a penalty.

The main goal of the agent is to maximize the total reward over time.

# Why Reinforcement Learning?

Some real-world problems do not have labeled data or predefined answers.

Instead of learning from examples, the machine learns by making decisions and observing the results.

Reinforcement Learning is useful when the machine needs to make a sequence of decisions to achieve a goal.

# Real-Life Example

Imagine teaching a small child to ride a bicycle.

✔ If the child rides correctly, you praise them.

✔ If they lose balance, they learn from the mistake.

After many attempts, the child learns how to ride the bicycle successfully.

Similarly,

Child → Agent

Road → Environment

Pedaling → Action

Praise → Reward

Mistake → Penalty

Learning → Reinforcement Learning

# Components of Reinforcement Learning

### 1. Agent

The learner or decision maker.

Example:
Robot, Self-driving Car

---

### 2. Environment

The surroundings where the agent works.

Example:
Road, Maze, Chess Board

---

### 3. State

Current situation of the agent.

Example:
Robot is at Position A.

---

### 4. Action

A decision taken by the agent.

Example:
Move Left
Move Right
Move Forward

---

### 5. Reward

Feedback received after performing an action.

Positive Reward → Good Action

Negative Reward → Bad Action

---

### 6. Policy

A strategy that tells the agent which action to take in each state.

# Workflow of Reinforcement Learning

Agent

↓

Observes Environment

↓

Chooses an Action

↓

Environment Responds

↓

Receives Reward

↓

Learns from Experience

↓

Repeats Until Goal is Achieved
# Example Scenario

Goal:

A robot wants to reach the destination.

If it reaches the goal,

Reward = +10

If it hits an obstacle,

Reward = -10

If it moves correctly,

Reward = +1

The robot keeps learning until it finds the best path.

```python
import pandas as pd

# Simple Reinforcement Learning Example

df = pd.DataFrame({

    "State":["Start","Middle","Obstacle","Goal"],

    "Action":["Move Right","Move Forward","Move Back","Stop"],

    "Reward":[1,2,-5,10]

})

df
```

```python
print("Agent Learning Process\n")

for i in range(len(df)):

    print("State :",df.loc[i,"State"])

    print("Action :",df.loc[i,"Action"])

    print("Reward :",df.loc[i,"Reward"])

    print()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

plt.bar(df["State"],df["Reward"])

plt.title("Rewards Received by Agent")

plt.xlabel("State")

plt.ylabel("Reward")

plt.grid()

plt.show()
```

# Difference Between Supervised, Unsupervised, and Reinforcement Learning

| Feature | Supervised Learning | Unsupervised Learning | Reinforcement Learning |
|----------|---------------------|-----------------------|------------------------|
| **Definition** | Learns from labeled data to predict outputs. | Learns from unlabeled data to discover hidden patterns. | Learns by interacting with the environment using rewards and penalties. |
| **Training Data** | Labeled Data | Unlabeled Data | Learns from experience |
| **Target Variable** | Available | Not Available | Not Available |
| **Main Objective** | Prediction | Pattern Discovery | Decision Making |
| **Output** | Predicted values or class labels | Clusters, associations, or reduced dimensions | Best action or policy |
| **Feedback** | Immediate | No Feedback | Reward or Penalty |
| **Learning Method** | Learns from known input-output pairs | Learns from similarities and patterns | Learns through trial and error |
| **Algorithms** | Linear Regression, Logistic Regression, Decision Tree, Random Forest, SVM, KNN, Naive Bayes | K-Means, DBSCAN, PCA, t-SNE, Apriori | Q-Learning, Deep Q-Network (DQN), SARSA |
| **Applications** | House Price Prediction, Spam Detection, Disease Prediction | Customer Segmentation, Recommendation Systems, Market Basket Analysis | Robotics, Self-Driving Cars, Game Playing, Resource Management |
