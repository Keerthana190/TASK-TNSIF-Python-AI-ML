##  Missing Values

Missing values are represented as:
- NaN (Not a Number)
- None

---

##  Steps:
1. Identify missing values
2. Count them
3. Decide:
   - Drop
   - Fill

##  Filling Missing Values

### Numerical Columns:
- Mean
- Median

### Categorical Columns:
- Mode

---

##  When to Use:
- Mean → Normal data
- Median → Skewed data
- Mode → Categories


###Part 2: Handling Missing Values
- Identify missing values Name,Age,Salary,Department,City,Joining_Date,Performance.
- Handle numerical missing values using appropriate techniques.
- Handle categorical missing values using appropriate techniques.
- Verify that all missing values have been handled successfully.

```python
#Step 1: Handle Numerical Columns

# Fill missing Age using Median
# (Median is preferred because the dataset contains outliers like 150 and -5)

df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill missing Salary using Median
# (Median is more robust because Salary contains an extreme outlier: 999999)

df["Salary"] = df["Salary"].fillna(df["Salary"].median())
```

```python
#Step 2: Handle Categorical Columns

## Fill missing Name using Mode
df['Name'] = df['Name'].fillna(df['Name'].mode()[0])

## Fill missing Department using Mode
df['Department'] = df['Department'].fillna(df['Department'].mode()[0])

## Fill missing City using Mode
df['City'] = df['City'].fillna(df['City'].mode()[0])

## Fill missing Performance using Mode
df['Performance'] = df['Performance'].fillna(df['Performance'].mode()[0])
```

```python
#Step 3: Handle Date Column

# Fill missing Joining_Date using Mode
# (Another common option is forward fill, but mode keeps this exercise simple.)

df["Joining_Date"] = df["Joining_Date"].fillna(df["Joining_Date"].mode()[0])
```

```python
print("Missing Values After Handling:\n")
print(df.isnull().sum())
```
