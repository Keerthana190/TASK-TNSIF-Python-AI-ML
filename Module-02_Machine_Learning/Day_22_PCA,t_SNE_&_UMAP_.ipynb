# What is Dimensionality Reduction?

## Definition

Dimensionality Reduction is an Unsupervised Machine Learning technique used to reduce the number of input features while preserving the most important information.

It helps simplify datasets, improve model performance, and reduce computational cost.

### Formula

High-Dimensional Data → Dimensionality Reduction Algorithm → Low-Dimensional Data

# Why Dimensionality Reduction?

- Reduces the number of features
- Removes redundant information
- Reduces training time
- Improves visualization
- Helps prevent overfitting
- Makes models easier to understand

# Types of Dimensionality Reduction

## 1. Principal Component Analysis (PCA)

A linear technique that transforms features into principal components while preserving maximum variance.

---

## 2. t-Distributed Stochastic Neighbor Embedding (t-SNE)

A non-linear technique mainly used for visualization of high-dimensional datasets.

---

## 3. UMAP

A non-linear dimensionality reduction technique that preserves both local and global data structures.

# Principal Component Analysis (PCA)

## Definition

PCA (Principal Component Analysis) is a linear dimensionality reduction algorithm.

It converts correlated features into a smaller number of uncorrelated principal components while preserving maximum information.

## Applications

- Image Compression
- Face Recognition
- Feature Reduction
- Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
data = {
    "Age":[25,35,45,55,65],
    "BP":[120,130,140,150,160],
    "Cholesterol":[180,200,220,240,260],
    "Sugar":[90,100,120,140,160],
    "BMI":[22,24,27,30,33]
}

df = pd.DataFrame(data)

print(df)
```

```python
pca = PCA(n_components=2)

result = pca.fit_transform(df)

print(result)
```

```python
plt.figure(figsize=(6,4))

plt.scatter(result[:,0], result[:,1])

plt.title("PCA Visualization")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)

plt.show()
```

# t-Distributed Stochastic Neighbor Embedding (t-SNE)

## Definition

t-SNE is a non-linear dimensionality reduction algorithm mainly used for data visualization.

It converts high-dimensional data into two or three dimensions while preserving the relationships between nearby data points.

## Applications

- Cluster Visualization
- Pattern Discovery
- Image Visualization

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(
    n_components=2,
    random_state=42,
    perplexity=2
)

result = tsne.fit_transform(df)

print(result)
```

```python
plt.figure(figsize=(6,4))

plt.scatter(result[:,0], result[:,1])

plt.title("t-SNE Visualization")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.grid(True)

plt.show()
```

# UMAP

## Definition

UMAP (Uniform Manifold Approximation and Projection) is a non-linear dimensionality reduction technique.

It preserves both local and global data structures while reducing dimensions.

## Applications

- Data Visualization
- Image Analysis
- Bioinformatics
- Pattern Recognition

# PCA vs t-SNE

| PCA | t-SNE |
|------|--------|
| Linear Technique | Non-linear Technique |
| Faster | Slower |
| Preserves Maximum Variance | Preserves Local Relationships |
| Used for Feature Reduction | Used for Visualization |
| Handles Large Datasets | Better for Small Datasets |

# Advantages

- Reduces the number of features
- Improves model performance
- Faster training
- Reduces computational cost
- Better data visualization

# Disadvantages

- Some information may be lost
- PCA works only for linear relationships
- t-SNE is computationally expensive
- Reduced features are harder to interpret
