#Mini Project:

Create a Student Report Card Program using:

1.List → store student names

2.Set → store subjects

3.Tuple → store student information (name, age)

4.Dictionary → store subject-wise marks

Requirements:

1.Ask the user to enter:

*Student name

*Age

*Number of subjects                                                                                                                                     2.For each subject:

*Enter subject name

*Enter marks

3.Store:

*Name in a list

*Subjects in a set

*Student info in a tuple

*Marks in a dictionary

4.Print a report card showing:

*Name

*Age

*Subjects

*Marks

*Total marks                                                                                                                                                   *Average marks  Actually this is the task she gave for datatypes so generate code for this and also tell me where to keep it

```python
# List to store student names
student_names = []

# Student Details
name = input("Enter Student Name: ")
age = int(input("Enter Age: "))
num_subjects = int(input("Enter Number of Subjects: "))

student_names.append(name)

# Set to store subjects
subjects = set()

# Tuple to store student information
student_info = (name, age)

# Dictionary to store marks
marks = {}

# Getting subject details
for i in range(num_subjects):
    print("\nEnter Subject", i + 1)

    subject = input("Subject Name: ")
    mark = int(input("Marks: "))

    subjects.add(subject)
    marks[subject] = mark

# Calculations
total_marks = sum(marks.values())
average_marks = total_marks / num_subjects

# Report Card
print("\n----- STUDENT REPORT CARD -----")

print("Name:", student_info[0])
print("Age:", student_info[1])

print("\nSubjects:", subjects)

print("\nMarks:")
for subject in marks:
    print(subject, ":", marks[subject])

print("\nTotal Marks:", total_marks)
print("Average Marks:", average_marks)
```

```python
student_names = []

subjects = set()

marks_dict = {}

name = input("Enter Student Name: ")
age = int(input("Enter Age: "))

student_names.append(name)

student_info = (name, age)

n = int(input("Enter Number of Subjects: "))

for i in range(n):
    subject = input("Enter Subject Name: ")
    marks = int(input("Enter Marks: "))

    subjects.add(subject)
    marks_dict[subject] = marks

total = sum(marks_dict.values())
average = total / n

print("\nREPORT CARD")
print("Name:", student_info[0])
print("Age:", student_info[1])
print("Subjects:", subjects)
print("Marks:", marks_dict)
print("Total:", total)
print("Average:", average)
```
