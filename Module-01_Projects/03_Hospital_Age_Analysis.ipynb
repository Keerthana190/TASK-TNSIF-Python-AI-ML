##Challenge
A hospital collected the ages of 500 patients. Visualize the age distribution and identify the most common age group.
Try it with Kaggle dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read Dataset
data = pd.read_csv("hospital data analysis.csv")

# Display first five rows
print(data.head())

# Histogram
plt.figure(figsize=(8,5))

plt.hist(data["Age"],
         bins=10,
         color="green",
         edgecolor="black")

plt.title("Hospital Patient Age Distribution")

plt.xlabel("Age")

plt.ylabel("Number of Patients")

plt.show()
```

```python
#Finding Most Common Age group

print(data["Age"].value_counts())
```

```python
most_common_age = data["Age"].mode()[0]

print("Most Common Age :", most_common_age)
```
