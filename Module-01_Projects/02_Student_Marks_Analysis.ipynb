##Problem : Student Marks Analysis

- Create a dataset containing marks of 10 students.

- students = ['A','B','C','D','E','F','G','H','I','J']
- marks = [78,85,92,67,88,95,72,81,90,76]
Tasks
1. Display student marks graphically.
2. Find the highest scorer.
3. Add appropriate title and axis labels.
4. Display marks in descending order and plot again.

```python
#1.Display student marks graphically

import matplotlib.pyplot as plt

# Student names
students = ['A','B','C','D','E','F','G','H','I','J']

# Student marks
marks = [78,85,92,67,88,95,72,81,90,76]

# Create Bar Graph
plt.figure(figsize=(8,5))

plt.bar(students, marks)

plt.title("Student Marks Analysis")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

```python
#2.Find the highest scorer.

highest_marks = max(marks)

index = marks.index(highest_marks)

highest_student = students[index]

print("Highest Student :", highest_student)
print("Highest Marks :", highest_marks)
```

```python
#3.Display marks in descending order

sorted_data = sorted(zip(marks,students), reverse=True)

print(sorted_data)
```

```python
#4.plot again

marks, students = zip(*sorted_data)

plt.figure(figsize=(8,5))

plt.bar(students, marks)

plt.title("Student Marks Analysis")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```
