```python
#Arithmetic Operators

'''Arithmetic operators are used to perform mathematical calculations'''

a = 40
b = 20

print("Addition:", a+b)
print("Subtraction:", a-b)
print("Mulitiplication:", a*b)
print("Division:", a/b)
print("Floor Division:", a//b)
print("Modulus:", a%b)
print("Power:", a**b)
```

```python
#Relational Operators

'''Comparison or Relational operators compare two values and return either True or False.'''

a = 25
b = 22

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

```python
#Assignment Operators

'''Assignment operators are used to assign values to variables and perform arithmetic operations while assigning.'''

x = 25

x += 5
print(x)

x -= 8
print(x)

x *= 2
print(x)

x /= 2
print(x)

x %= 2
print(x)

x //= 2
print(x)

x **= 2
print(x)
```

```python
#Logical Operators

'''Logical operators are used to combine two or more conditions.'''

a = True
b = False

print(a and b)
print(a or b)
print(not a)
print(not b)

c = 10 > 20
d = 20 > 10
print(c and d)
print(c or d)
print(not c)
print(not d)
```

```python
#Membership operators

'''Membership operators check whether an element exists in a sequence or collection.'''

#Example-1
numbers = [10,20,30,40]

print(20 in numbers)
print(50 not in numbers)


#Example-2
student = {
    "name" : "Keerthana",
    "age" : 22
}

print("name" in student)   # Use quotes for strings (dictionary keys), not for variables.
print("age" in student)
print("name" not in student)
print("age" not in student)

#Example-3
text = "Python"

print("P" in text)
print("Java" in text)

#Example-4
colors = {"Red", "Blue", "Green"}

print("Blue" in colors)
```

```python
#Identity Operators

'''Identity operators check whether two variables refer to the same object in memory.'''
a = [1,2,3]
b = a
c = [1,2,3]

print(a is b)   # 'is' checks whether both variables refer to the same object in memory.
print(a is c)

print(a == c)   # '==' compares the values of two objects.
```

```python
#Bitwise Operators

'''Bitwise operators perform operations on the binary representation of integers.'''

a = 5
b = 3

print(a & b)
print(a | b)
print(a ^ b)
print(~a)
print(a << 1)
print(a >> 1)
```

###2.Conditional Statements
Conditional statements execute different blocks of code based on whether a condition is True or False.


if Statement :
Executes a block of code only if the condition is True.

Syntax:

if condn : stmt

```python
#Program1 to check the age
Age1 = 20

if Age1 >= 18:
  print("ELigible to vote")
print("program executed successfully")

#Program2 to check the age
Age2 = int(input("Enter Age: "))

if Age2 >= 18:
  print("Eligible to vote")
print("program executed successfully")
```

if-else Statement:
Executes one block if the condition is True, otherwise executes another block.

Syntax:

if condn :
  stmt
else:
  stmt

```python
#Program1 to check the age
Age2 = int(input("Enter Age: "))

if Age2 >= 18:
  print("Eligible to vote")
else:
  print("Not Eligible to vote")
print("program executed successfully")
```

```python
#Program2 to check whether a number is even or odd

Number = 5

if Number % 2 == 0:
  print("Even Number")
else:
  print("Odd Number")
```

if-elif-else Statement:
Checks multiple conditions one by one until one becomes True.

```python
#Program1 to check student1 grades

marks = int(input("Enter Marks:"))

if marks >= 90:
  print("Grade A")
elif marks >= 50:
  print("Grade B")
elif marks >= 40:
  print("Grade C")
else:
  print("Fail")

##Program2 to check student2 grades
marks = int(input("Enter marks"))
if(marks<=100 and marks>=80):
  print("A grade")
elif (marks<80 and marks>=50):
  print("B grade")
elif(marks<50 and marks>=40) :
  print("C grade")
elif(marks<40 and marks>=0) :
  print("fail")
else :
  print("Enter valid input")
```

Nested if:
An if statement inside another if statement.

```python
#Program1 to check voting eligiblity using nested if
age = 22
citizen = True

if age >= 18:
    if citizen:
        print("Eligible")

#Program2 to check voting eligiblity using nested if else
age = int(input("Enter your age"))
citizen = True
if(citizen) :
  if(age>=18) :
    print("Eligible")
  else:
    print("not eligible")
```

##3.Control Statements:
Control statements change the normal flow of program execution.

###For loop
Used to repeat a block of code for a fixed number of iterations or over a sequence.

####syntax :
for i in range(): stmts

```python
#Program to Print numbers from 1 to 5
for i in range(6):  #Stop value
  print(i)
```

```python
for i in range(1,6): #Start and Stop values
    print(i)
```

```python
for i in range(3,51,3):  #Start, Stop and Step values
  print(i)
```

While Loop:

Repeats a block of code as long as the condition remains True.


```python
#Program to Print first 10 natural numbers
num=1
while num<=10:
  print(num)
  num+=1


#Program to Print first n natural numbers
number = int(input("Enter a number: "))

count = 1
while count <= number:
    print(count)
    count += 1
```

###break Statement:
Terminates the loop immediately.

```python
for i in range(1,10):

    # Stop loop when i becomes 6
    if i == 6:
        break

    print(i)
```

###continue Statement:
Skips the current iteration and moves to the next iteration.

```python
for i in range(1,6):

    # Skip printing 3
    if i == 3:
        continue

    print(i)
```

###pass Statement:
pass is a placeholder statement. It does nothing and is used when a statement is syntactically required.

```python
for i in range(5):

    # Placeholder for future implementation
    if i == 2:
        pass

    print(i)
```

```python
# 3.Reverse of a number

#Extraction of numbers in reverse order = Loops(Control statements)

number = int(input("Enter a number: "))

reverse = 0

while number > 0:
  remainder = number % 10
  reverse = reverse *10 + remainder
  number = number // 10

print("Reverse of the number is: ", reverse)
```

```python
# 4.Sum of the digits

number = int(input("Enter a number: "))

sum = 0

while number > 0:
  remainder = number % 10
  sum = sum + remainder
  number = number // 10

print("Sum of the digits is: ", sum)
```

```python
# 5.Palindrome number

# Extraction and Condition check = Conditional and control statements

number = int(input("Enter a number: "))

reverse = 0
temp = number

while number > 0:
  remainder = number % 10
  reverse = reverse * 10 + remainder
  number = number // 10

if reverse == temp:
  print("Given number is palindrome")
else:
  print("Given number is not a palindrome")
```

```python
# 6.Armstrong number
# Extraction and Condition check

#While loop
number = int(input("Enter a number: "))

sum = 0
count = len(str(number))
temp = number

while number > 0:
  remainder = number % 10
  sum = sum + (remainder ** count)
  number = number // 10

if sum == temp:
  print("Given number is armstrong number")
else:
  print("Given number is not an armstrong number")


#For loop
number = int(input("Enter a number: "))

sum = 0
temp = number
count = len(str(number))

for i in str(number):
  sum = sum + (int(i) ** count)

if sum == temp:
  print("Given number is armstrong number")
else:
  print("Given number is not an armstrong number")
```

```python
# 8.Factorial of a number

#for loop
number = int(input("Enter a number: "))

fact = 1

for i in range(1,number+1):
  fact = fact * i

print(fact)


#while loop
num = int(input("Enter a number: "))

fact = 1
i = 1
while i <= num:
  fact = fact * i
  i += 1

print(fact)
```

```python
# 9.Fibonacci series

number = int(input("Enter a number:"))

a = 0
b = 1
print(a)
print(b)

for i in range(number):
  c = a + b
  print(c)
  a = b
  b = c
```

```python
# 10.Reverse of a string

#String Handling problems

string = input("Enter a string:")

reverse = ""
for i in string:
  reverse = i + reverse

print(reverse)


#Palindrome
string = input("Enter a string:")

reverse = ""
for i in string:
  reverse = i + reverse

if string == reverse:
  print("Given string is palindrome")
else:
  print("Given string is not a palindrome")
```

```python
# 11.Patterns

#Square

print("===Square===")
for i in range(5):
  for j in range(5):
    print("* ", end='')
  print()


#Right Angled Triangle

print("===Right Angled Triangle===")
for row in range(6):
  for col in range(row):
    print("* ", end='')
  print()

#Inverted Triangle

print("===Inverted Triangle===")
for row in range(6,0,-1):
  for col in range(row):
    print("* ", end='')
  print()

#Number Triangle

print("===Number Triangle===")
for row in range(6):
  for col in range(1,row+1):
    print(col, end='')
  print()

#Same Number Triangle

print("===Same Number Triangle===")
for row in range(6):
  for col in range(row):
    print(row, end='')
  print()
```
