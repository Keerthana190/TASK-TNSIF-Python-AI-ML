#Introduction to Matplotlib

Matplotlib is one of the most popular Python libraries used for data visualization. It helps us represent data graphically so that trends, patterns, and comparisons become easier to understand.

##Why we use Matplotlib?
- Easy to create graphs and charts
- Helps in data analysis
- Used in Data Science, Machine Learning, Business Analytics, and Research
- Supports various types of visualizations

##Installation

- Matplotlib is already available in Google Colab.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## List of Functions and Their Usage

| Function | Usage |
|----------|--------|
| `plt.plot(x,y)` | Line plot |
| `plt.bar(x,y)` | Bar chart |
| `plt.hist(data, bins)` | Histogram |
| `plt.scatter(x,y)` | Scatter plot |
| `plt.pie(values, labels)` | Pie chart |
| `plt.xlabel(), plt.ylabel()` | Set axis labels |
| `plt.title()` | Set plot title |
| `plt.legend()` | Add legend |
| `plt.grid()` | Show grid |
| `plt.subplot()` | Multiple plots in one figure |
| `plt.figure(figsize=(w,h))` | Define figure size |
| `plt.show()` | Display plot |

##Basic Plot Structure




```python
import matplotlib.pyplot as plt

Marks = [75, 80, 83, 85, 95, 90]
Amount = [2500, 3800, 4200, 4500, 5500, 5000]
plt.plot(Marks, Amount)
plt.xlabel('Marks')
plt.ylabel('Amount')
plt.title('Vijay Scholarship')
plt.legend('Scholar')
plt.show()
```

##When to Use?

When you want to show trends over time.

##Applications
- Stock prices
- Temperature changes
- Sales growth
- Population growth
##Best Use Case

Showing continuous data over time.

##Bar Chart
- Definition

Represents data using rectangular bars.

- When to Use?

When comparing different categories.

- Applications
1. Student marks comparison
2. Product sales comparison
3. Department performance

- Best Use Case

Comparing multiple categories.

```python
students = ['A','B','C','D']
marks = [85,90,78,95]

plt.bar(students, marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt

departments = ["CSE", "ECE", "EEE", "MECH", "CIVIL"]
students = [180, 120, 90, 70, 60]

plt.bar(departments, students)

plt.title("Number of Students in Each Department")
plt.xlabel("Department")
plt.ylabel("Students")

plt.show()
```

```python
import matplotlib.pyplot as plt

brands = ["HP","Dell","Lenovo","Acer","Asus"]
sales = [250,300,180,150,200]

plt.bar(brands, sales)

plt.title("Laptop Sales")
plt.xlabel("Brand")
plt.ylabel("Units Sold")

plt.show()
```

##Histogram
- Definition

Shows frequency distribution of numerical data.

- When to Use?

To understand data distribution.

- Applications
1. Exam score analysis
2. Salary distribution
3. Age distribution

- Best Use Case

Understanding how data is spread.

```python
import matplotlib.pyplot as plt

marks = [35,42,50,55,60,62,65,68,70,72,75,78,80,82,85,88,90,92,95]

plt.hist(marks, bins=5)

plt.title("Distribution of Student Marks")
plt.xlabel("Marks")
plt.ylabel("Number of Students")

plt.show()
```

```python
import matplotlib.pyplot as plt

salary = [25000,27000,30000,32000,35000,
          36000,38000,42000,45000,47000]

plt.hist(salary, bins=5)

plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Employees")

plt.show()
```

```python
ages = [18000,19000,20000,20000,21000,22000,22000,23000,24000,25000,25000,25000]

plt.hist(ages, bins=5)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

##Scatter Plot
- Definition

Represents individual data points.

- When to Use?

To find relationships between variables.

- Applications
1. Height vs Weight
2. Study Hours vs Marks
3. Experience vs Salary
- Best Use Case

Finding correlations between variables.

```python
import matplotlib.pyplot as plt

hours = [1,2,3,4,5,6,7,8]
marks = [40,48,55,63,72,80,88,95]

plt.scatter(hours, marks)

plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.show()
```

```python
import matplotlib.pyplot as plt

hours = [1,2,3,4,5,6]
calories = [200,350,480,600,720,850]

plt.scatter(hours, calories)

plt.title("Exercise vs Calories Burned")
plt.xlabel("Hours Exercised")
plt.ylabel("Calories")

plt.show()
```

##Pie Chart
- Definition

Represents data as portions of a whole.

- When to Use?

To show percentage contributions.

- Applications
1. Market share
2. Budget allocation
3. Survey results
- Best Use Case

Showing percentages.

```python
import matplotlib.pyplot as plt

expense = [15000,5000,3000,4000]
labels = ["Rent","Food","Travel","Shopping"]

plt.pie(expense,
        labels=labels,
        autopct="%1.1f%%")

plt.title("Monthly Expenses")

plt.show()
```

```python
labels = ['Python','Java','C++','JavaScript']
values = [40,25,20,15]

plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title("Programming Language Usage")
plt.show()
```

##Multiple Plots in One Figure

Subplots (plt.subplot())
- Definition

Displays multiple graphs in one figure for comparison.

- When to Use?

Comparing related datasets side by side.

- Real-life Examples
- Sales vs Profit
- Temperature vs Rainfall
- Student Marks vs Attendance
- Revenue vs Expenses

```python
x = [1,2,3,4,5]

plt.subplot(1,2,1)
plt.plot(x,[1,4,9,16,25])
plt.title("Plot 1")

plt.subplot(1,2,2)
plt.plot(x,[1,2,3,4,5])
plt.title("Plot 2")

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.subplot(1,2,1)
plt.plot([1,2,3],[2,4,6])
plt.title("Line Plot")

plt.subplot(1,2,2)
plt.bar(["A","B","C"],[5,7,3])
plt.title("Bar Plot")

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.subplot(1,2,1)
plt.bar(["A","B","C"], [10,20,15])
plt.title("Sales")

plt.subplot(1,2,2)
plt.plot([1,2,3],[30,25,35])
plt.title("Profit")

plt.show()
```

```python
plt.plot([10,20,30],[5,8,6])

plt.grid(True)

plt.show()
```

##Graph Customization

```python
#Change Line Color
#Definition: Changes the color of the plotted line.

plt.plot([1,2,3],[4,5,6], color='pink')
plt.show()
```

```python
#Change Line Style
#Definition: Changes the appearance of the line.

plt.plot([1,2,3],[4,5,6], linestyle='--')
plt.show()
```

```python
#Change Marker
#Definition: Displays a symbol at every data point.

#Change Marker
plt.plot([1,2,3],[4,5,6], marker='o')
plt.grid(True)
plt.show()
```

```python
#Add Grid
plt.grid(True)
```

```python
#plt.figure(figsize=(w, h)) is used to set the size of the graph (figure).

x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.figure(figsize=(8, 4))
plt.plot(x, y)

plt.title("Line Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [15,25,20,35,30]

plt.figure(figsize=(8,4))

plt.plot(
    x,
    y,
    color="purple",
    linestyle="--",
    marker="o",
    linewidth=2
)

plt.title("Customized Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.grid(True)

plt.show()
```
