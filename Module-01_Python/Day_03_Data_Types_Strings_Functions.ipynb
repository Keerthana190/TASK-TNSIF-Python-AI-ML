#1. List
Definition

A List is an ordered, mutable (changeable) collection of items. It allows duplicate values and can store different data types.

Features
Ordered
Mutable
Allows duplicates
Supports indexing

```python
#Example

fruits = ["apple", "banana", "mango"]
print(fruits)
print(type(fruits))
print(fruits[0])
```

Important Functions
Function	Description
- append()	Add item at end
- insert()	Insert item at specific position
- remove()	Remove item
- pop()	Remove item using index
- sort()	Sort list
- reverse()	Reverse list
- len()	Count elements
- count()	Count occurrences

```python
#List-Example-1 - Student Marks List

marks = [85, 90, 78, 95, 90]
print(marks)
print(type(marks)) #checks type of data
marks.append(55)   #Adds an element
print(marks)
marks.insert(2,65) #Add an element at a particular index
print(marks)
marks.remove(55)   #remove an element
print(marks)
marks.pop(2)       #remove an element using index
print(marks)
marks.sort()       #sorts the elements
print(marks)
marks.reverse()    #prints elements in reverse order
print(marks)
print(len(marks))  #prints length of list
marks.count(90)    #count the occurences of an element
```

```python
#List-Example-2 - Student Names List

stud = ['nikhitha','kiran','madhu']
print(stud)
print(type(stud))
stud.append("shila")
print(stud)
stud.insert(1,'naresh')
print(stud)
stud.remove('madhu')
print(stud)
stud.pop(0)
print(stud)
stud.sort()
print(stud)
stud.reverse()
print(stud)
l = len(stud)
print(l)
fruit = "Banana"
fruit.count('a')
```

```python
#List-Example-3 = Nick names

stu = ['sunny', 'bunny', 'kanni', 'ramu', 'sai']
print(stu)
print(type(stu))
stu.append('keerthi')
print(stu)
stu.insert(2,'saba')
print(stu)
stu.remove('bunny')
print(stu)
stu.pop(3)
print(stu)
stu.sort()
print(stu)
stu.reverse()
print(stu)
print(stu.index('sunny'))
stu.pop()
print(stu)
stu.clear()
print(stu)
stu.copy()
print(stu)
stu.append('keerthi')
print(stu)
stu.copy()
print(stu)
stu.extend('keerthi')
print(stu)
stu.count('keerthi')
print(stu)
```

#2. Tuple
Definition

A Tuple is an ordered and immutable (cannot be changed) collection of items.

Features
Ordered
Immutable
Allows duplicates
Faster than lists



```python
#Example

colors = ("Red", "Green", "Blue")

print(colors)
print(type(colors))
print(colors[1])
```

Important Functions &
Function	Description
- len()	Number of elements
- count()	Count occurrence
- index()	Find position


```python
#Tuple-Example-1 - Student Semester SGPAs
sgpa = (8.2, 8.5, 8.8, 9.0, 8.5, 9.2)
print(sgpa)
print(len(sgpa))
print(sgpa.count(8.5))
print(sgpa.index(9.0))
```

#3. Set
Definition

A Set is an unordered collection of unique elements.

Features
Unordered
Mutable
No duplicates allowed
Faster searching

```python
#Example

numbers = {1, 2, 3, 4}

print(numbers)
print(type(numbers))
```

Important Functions &
Function	Description
- add()	Add element
- remove()	Remove element
- union()	Combine sets
- intersection()	Common elements
- difference()	Unique elements
- clear()	Remove all elements

```python
#Set-Example-1 - two sets of names
set1 = {'srija','shiva','teja','priya','Ajith'}
set2 = {'teja','priya','Ajith','kiran','ashok'}
print(set1)
print(set2)
set1.add('keerthi')
print(set1)
set2.remove('priya')
print(set2)
print(set1.union(set2))
union = set1.union(set2)
print(union)
print(set1.intersection(set2))
inter = set1.intersection(set2)
print(inter)
print(set1.difference(set2))
print(set2.difference(set1))
diff = set1.difference(set2)
print(diff)
set1.clear()
print(set1)
```

```python
#Set-Example-2 - Registered and Selected student sets
Registered_students = {"Saba", "Charan", "Keerthi", "Pavani", "Nikki", "Abhishek", "Ravi", "Kiran"}
Selected_students = {"Saba", "Charan", "Pavani", "Keerthi"}
print(Registered_students)
print(Selected_students)
Registered_students.add("Siri")
print(Registered_students)
Registered_students.remove("Kiran")
print(Registered_students)
print(Registered_students.union(Selected_students))
print(Registered_students.intersection(Selected_students))
print(Registered_students.difference(Selected_students))
print(Selected_students.difference(Registered_students))
Registered_students.clear()
print(Registered_students)
```

#4. Dictionary
Definition

A Dictionary stores data in key-value pairs.

Features
Mutable
Keys must be unique
Fast lookup
Stores related information together

Important Functions &
Function	Description
- keys()	Returns all keys
- values()	Returns all values
- items()	Returns key-value pairs
- get()	Safely access value through key
- pop()	Remove key
- update()	Update dictionary


```python
#Example-1 -Car Brand Details

car_brands = {
    "brand": "Toyota",
    "model": "Corolla",
    "year": 2021
}
print(car_brands)
print(car_brands.keys())
print(car_brands.values())
print(car_brands.items())
print(car_brands.get('brand'))
print(car_brands.pop('model'))
print(car_brands)
print(car_brands.update({'model': 'Corolla'}))
print(car_brands)
print(car_brands.update({'year':'2022'}))
print(car_brands)
```

```python
#Example-2 -Student Details
student = {
    "name": "Keerthana",
    "age": 22,
    "course": "AIML",
    "Gender": "Female",
    "Address": "Hyd"
}
print(student)
print(student.keys())
print(student.values())
print(student.items())
print(student.get('name'))
student.pop('name')
print(student)
student.update({'name':'Saba'})
print(student)
student.update({'age':21})
print(student)
```

#Strings in Python
What is a String?

A String is a sequence of characters enclosed within single quotes (' '), double quotes (" "), or triple quotes (''' ''' or """ """).

Features of Strings
- Strings are ordered
- Strings are immutable (cannot be modified directly)
- Supports indexing and slicing
- Can contain letters, numbers, and special characters

##String Indexing

Each character has a position called an index.

- text = "Python"
- Character	P	y	t	h	o	n
- Positive Index	0	1	2	3	4	5
- Negative Index	-6	-5	-4	-3	-2	-1

Functions and purpose
- upper()	Convert to uppercase
- lower()	Convert to lowercase
- capitalize()	First letter uppercase
- title()	Every word first letter uppercase
- strip()	Remove spaces
- replace('replaceval)	Replace text
- split()	Convert string to list
- find('str')	Find position returns index
- count('char)	Count occurrences
- startswith(value)	Check starting text returns bool
- endswith('value)	Check ending text returns bool

```python
#Example-1 - Course Title
Course_name = 'python and AI & ML'
print(Course_name)
print(len(Course_name))
print(Course_name.upper())
print(Course_name.lower())
print(Course_name.capitalize())
print(Course_name.title())
print(Course_name.replace('python','java'))
print(Course_name.split())
print(Course_name.find('AI'))
print(Course_name.count('n'))
print(Course_name.startswith("Python"))
print(Course_name.endswith("programming"))
print(Course_name.endswith("ML"))
```

```python
#Example-2 -Greet string
Greet = "_ welcome to Artificial Intelligence and Machine Learning ___"
print(Greet)
print(Greet.upper())
print(Greet.lower())
print(Greet.capitalize())
print(Greet.title())
print(Greet.strip())
print(Greet.replace('and','&'))
print(Greet.split())
print(Greet.find('Machine'))
print(Greet.count('a'))
print(Greet.startswith("_"))
print(Greet.endswith("Learning"))
print(Greet)
```

#Functions in Python
What is a Function?

A function is a block of reusable code that performs a specific task. Instead of writing the same code multiple times, we can create a function and call it whenever needed.

Benefits of Functions
Code Reusability
Reduces Code Duplication
Improves Readability
Easier Debugging and Maintenance

```python
#Functions
#Normal function without parameters and without return values

def greet():
  print("Hello Keerthana")

greet()
```

```python
#Function with parameter and without return value

def greet(name):
  print("Hello", name)

greet("Keerthana")
```

```python
#Function with Parameters and Return Value (Normal Inputs)

def add(num1, num2):
    return num1 + num2

result = add(10, 20)
print(result)
```

```python
#Function with Parameters and Return Value (Runtime Inputs)

def add(a, b):
    return a + b

A = int(input("Enter first number: "))
B = int(input("Enter second number: "))

print(add(A, B))
```

```python
#1 Function without Parameters but with Return Value

def add():
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))

    return num1 + num2

print(add())
```

```python
#2 Function without parameters and with return values

def display():
  name = input("Enter your name: ")
  age = int(input("Enter your age: "))
  return name
  return age

print(display())
```

```python
#Function without parameters and without return values but with runtime inputs and print statements

def display():

    name = input("Enter your name: ")
    age = int(input("Enter your age: "))

    print("Hello", name)
    print("Your age is", age)

display()
```

```python
#Arithmetic Calculator using Functions

def add(a,b):
    return a+b

def sub(a,b):
    return a-b

def mul(a,b):
    return a*b

def div(a,b):
    return a//b

A = int(input("Enter first number: "))
B = int(input("Enter second number: "))

print("Addition =", add(A,B))
print("Subtraction =", sub(A,B))
print("Multiplication =", mul(A,B))
print("Division =", div(A,B))
```

```python
'''Factorial program with parameters'''
def factorial(num):

  fact = 1
  for i in range(1,num+1):
    fact = fact * i

  return fact

print(factorial(5))

'''With parameters , return values & runtime inputs'''
def factorial(num):

  fact = 1
  for i in range(1,num+1):
    fact = fact * i
  return fact

number = int(input("Enter the number: "))
print(factorial(number))

'''With parameters , using print stmts & runtime inputs'''

def factorial(num):

  fact = 1
  for i in range(1,num+1):
    fact = fact * i
  print(fact)

number = int(input("Enter the number: "))
factorial(number)
```

```python
def factorial(n):
  if n == 1:
    return 1
  else:
    return n *factorial(n-1)

print(factorial(6))


def factorial(n):
  if n == 1:
    return 1
  else:
    return n *factorial(n-1)

number = int(input())
print(factorial(number))
```
