```python
import numpy as np
import pandas as pd
```

```python
df = pd.Series([10,30,50,65,34])
print(df)
```

# Descriptive Statistics

**s.describe():** Provides a quick summary of the data.

This method gives a statistical summary of the Series, including count, mean, standard deviation, minimum, maximum, and quartile values.

```python
# Descriptive statistics
print(df.describe())
```

 **s.mean():** Computes the mean of the data.

 **s.std():** Computes the standard deviation.

 **s.min() and s.max():** Computes the minimum and maximum values.

```python
print("Mean : ",df.mean())
print("Standard_Deviation : " ,df.std())
print("Minimum Value : ", df.min())
print("Maximum Value", df.max())
```

#Data Manipulation

**s.map(func):** Applies a function to each element in the Series.

```python
#Map Function

map_func = df.map(lambda x :x * 5)
print(map_func)
print(df)
```

**s.apply(func):** Similar to map, but more flexible. (Can be used Data Frames as well, where as map is only for Series)

```python
applies = df.apply(lambda x : x * 5)
print(applies)
print(df)
```

**s.sort_values():** Sorts the Series.

```python
s1 = pd.Series([4,7,54,23,92,41])
sort_func = s1.sort_values()
print(sort_func)
```

**s.drop(index):** Drops specified labels from the Series.(not deletes)

```python
drop_func = df.drop(1)
print(drop_func)
print(df)
```

#Handling Missing Data

**s.isnull():** Checks for missing values, returns a Series of booleans.

```python
#Handling Missing Data

s2 = pd.Series([2,4,6,np.nan,7,0])
print(s2.isnull())
```

**s.notnull():** Opposite of isnull().

```python
print(s2.notnull())
```

**s.fillna(value):** Fills missing values with a specified value.

```python
print(s2.fillna(0))
```

**s.dropna():** Drops all rows that contain missing values.like None

```python
s3 = pd.Series([4,7,3,None,8,2,np.nan])
print(s3)
print(s3.dropna())
```

# Indexing, Slicing, and Filtering
**s.iloc[ ]:** Purely integer-location based indexing.
(+)-left to right,(-)-right to left

```python
#Indexing, Slicing and Filtering

s3 = pd.Series([1,3,7,4,9,6,2])
print(s3.iloc[1])
print(s3.iloc[-4])
print(s3.iloc[5])
```

**s.loc[ ]:** Label-based indexing.(only we can give positive values not negitive)

```python
print(s3.loc[1])
print(s3.loc[4])
print(s3.loc[5])
```

**s[s > n]:** Filters and returns elements greater than n.

```python
filter = s3[s3<5]
print(filter)
```

#Aggregation

##s.sum(): Sums up the values

```python
#Aggregation

s4 = pd.Series([2,4,6,8,3,5,7])
print(s4.sum())
```

##s.cumsum() : cumulative Sum

```python
print(s4.cumsum())
```

**s.aggregate(func):** Aggregates using one or more operations. like
s.aggregate(['sum','mean','std'])

```python
print(s4.aggregate(['sum','min','max','std','sum']))
```

##Importing Dataset
- Importing datasets into Pandas is straightforward, and Pandas supports various file formats like csv, xlsx, json, sql etc.
- df = pd.read_csv('')

df.shape: The df.shape attribute of a DataFrame returns a tuple representing the dimensionality of the DataFrame. The first element of the tuple is the number of rows, and the second is the number of columns. This is useful when you need to know how large the dataset is, such as when you are preprocessing data or ensuring that data manipulations have executed correctly.


df.columns: Lists all the column names in the DataFrame
- Display the data types of each column
- Display summary statistics for numerical columns

```python
# Step 1: Import required libraries
import pandas as pd
import numpy as np
# Step 2: Upload CSV file from your system
from google.colab import files
uploaded = files.upload()

# Step 3: Read the uploaded CSV file
# Replace 'your_file.csv' with actual uploaded file name
df = pd.read_csv(list(uploaded.keys())[0])

# Display first 5 rows
print("First 5 rows:")
print(df.head())

# Step 4: Basic Information
print("\nDataset Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())

# Step 5: Handling Missing Values
print("\nMissing Values:")
print(df.isnull().sum())

# Fill missing values with 0 (example)
df = df.fillna(0)

# Step 6: Filtering Data (example)
# Filter days with high temperature (>30)
if 'temperature' in df.columns:
    filtered_df = df[df['temperature'] > 10]
    print("\nFiltered Data (Temperature > 30):")
    print(filtered_df)

# Step 7: Adding New Column
# Example: classify weather based on temperature
if 'temperature' in df.columns:
    df['weather_type'] = df['temperature'].apply(
        lambda x: 'Hot' if x >= 30 else 'Normal'
    )

# Step 8: GroupBy Example
# Average temperature based on weather condition
if 'weather_condition' in df.columns:
    group_data = df.groupby('weather_condition')['temperature'].mean()
    print("\nAverage Temperature by Weather Condition:")
    print(group_data)

# Step 9: Save Modified File
df.to_csv("processed_data.csv", index=False)

# Step 10: Download the processed file
from google.colab import files
files.download("processed_data.csv")
```
