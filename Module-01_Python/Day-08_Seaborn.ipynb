## What is Seaborn?

Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive, informative, and statistical graphics with less code.

## Why Seaborn?
- Easy to create beautiful visualizations.
- Works directly with Pandas DataFrames.
- Provides built-in themes and color palettes.
- Supports statistical plotting.
- Requires less code compared to Matplotlib.

##Difference Between Matplotlib and Seaborn

| Feature | Matplotlib | Seaborn |
|----------|--------|----------|
| Level | Low-level visualization library | High-level visualization library |
| Built On | Base plotting library | Built on Matplotlib |
| Syntax | More code required | Less code required |
| Statistical Plots | Limited support | Excellent support |
| DataFrame Support | Requires manual handling | Directly works with Pandas DataFrames |
| Customization | Highly customizable | Easy customization |
| Learning Curve | Slightly difficult | Beginner-friendly |



```python
#Matplotlib Example
import matplotlib.pyplot as plt

students = ['A','B','C','D']
marks = [78,85,92,88]

plt.bar(students, marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()



#Seaborn Example
import seaborn as sns
import pandas as pd

df = pd.DataFrame({
    'Students':['A','B','C','D'],
    'Marks':[78,85,92,88]
})

sns.barplot(x='Students', y='Marks', data=df)
plt.title("Student Marks")
plt.show()
```

##Advantages and Disadvantages of Seaborn

1. Advantages
--------------
- Easy to use
- Attractive Visualizations
- Statistical Visualization
- DataFrame Integration (directly works with pandas)
- Built-in Color Palettes
- Less Coding
2. Disadvantages
------------------
- Less Flexibility -
For highly customized plots, Matplotlib is preferred.
- Slower for Large Datasets- May be slower than pure Matplotlib.
- Limited Low-Level Control-Fine-grained customization often requires Matplotlib.
- Dependency on Matplotlib-Seaborn cannot work independently.

## 3. List of Commonly Used Functions in Seaborn

1. `sns.scatterplot()` – Creates scatter plots.  
2. `sns.lineplot()` – Creates line plots.  
3. `sns.barplot()` – Shows average values with confidence intervals.  
4. `sns.countplot()` – Counts frequency of categorical data.  
5. `sns.boxplot()` – Creates box-and-whisker plots.  
6. `sns.violinplot()` – Combines boxplot and KDE for distribution.  
7. `sns.heatmap()` – Creates heatmaps for correlation matrices or tabular data.  
8. `sns.pairplot()` – Pairwise relationships in dataset.  
9. `sns.distplot()` / `sns.histplot()` – Distribution of variables.  
10. `sns.regplot()` – Regression line with scatter plot.  
11. `sns.catplot()` – Plotting categorical data.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "Hours": [1, 2, 3, 4, 5, 6],
    "Marks": [35, 45, 55, 65, 78, 90]
}

df = pd.DataFrame(data)

sns.scatterplot(x="Hours", y="Marks", data=df)

plt.title("Hours Studied vs Marks")
plt.show()
```

```python
#1.Scatter Plot (sns.scatterplot())

df = pd.DataFrame({
'height' : [120,150,180,170,125,173,183],
'weight' : [50,45,62,71,65,83,92]
      })
sns.scatterplot(data=df)
plt.show()
```

```python
#2.Line Plot (sns.lineplot())

days = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
temp = [30,31,33,32,35,36,34]

df = pd.DataFrame({
    "Day": days,
    "Temperature": temp
})

sns.lineplot(x="Day", y="Temperature", data=df, marker="o")

plt.title("Weekly Temperature")
plt.show()
```

```python
#3.Bar Plot (sns.barplot())

department = ["HR","IT","Sales","Finance"]
salary = [35000,60000,45000,50000]

df = pd.DataFrame({
    "Department": department,
    "Salary": salary
})

sns.barplot(x="Department", y="Salary", data=df)

plt.title("Average Salary")
plt.show()
```

```python
#4.Count Plot (sns.countplot())

language = ["Python","Java","Python","C","Java","Python","C"]

df = pd.DataFrame({
    "Language": language
})

sns.countplot(x="Language", data=df)

plt.title("Programming Language Preference")
plt.show()
```

```python
#5.Box Plot (sns.boxplot())

salary = [25000,27000,30000,35000,36000,38000,39000,120000]

df = pd.DataFrame({
    "Salary": salary
})

sns.boxplot(y="Salary", data=df)

plt.title("Employee Salary Distribution")
plt.show()
```

```python
#6.Violin Plot (sns.violinplot())

marks = [45,55,60,62,65,67,70,75,80,85,90]

df = pd.DataFrame({
    "Marks": marks
})

sns.violinplot(y="Marks", data=df)

plt.title("Distribution of Marks")
plt.show()
```

```python
#7. Heatmap (sns.heatmap())

marks = pd.DataFrame(
{
    "Math":[80,70,90],
    "Science":[85,65,95],
    "English":[75,88,92]
},
index=["A","B","C"]
)

sns.heatmap(marks,
            annot=True,
            cmap="YlGnBu")

plt.title("Student Marks Heatmap")
plt.show()
```

```python
#8. Pair Plot (sns.pairplot())

data = {
    "Math":[80,70,90,85],
    "Science":[75,68,95,88],
    "English":[78,72,91,84]
}

df = pd.DataFrame(data)

sns.pairplot(df)

plt.show()
```

```python
#9. Histogram (sns.histplot())

marks = [45,55,60,65,70,72,75,80,82,85,90,92]

df = pd.DataFrame({
    "Marks": marks
})

sns.histplot(df["Marks"], bins=6)

plt.title("Marks Distribution")
plt.show()
```

```python
#10. Regression Plot (sns.regplot())

hours = [1,2,3,4,5,6]
marks = [35,42,58,66,80,91]

df = pd.DataFrame({
    "Hours": hours,
    "Marks": marks
})

sns.regplot(x="Hours", y="Marks", data=df)

plt.title("Study Hours vs Marks")
plt.show()
```

```python
#11. Cat Plot (sns.catplot())

city = ["Hyderabad","Delhi","Mumbai","Chennai"]
sales = [120,150,140,135]

df = pd.DataFrame({
    "City": city,
    "Sales": sales
})

sns.catplot(
    x="City",
    y="Sales",
    data=df,
    kind="bar"
)

plt.show()
```

```python
# Seaborn Implementation - All in One Code

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
tips = sns.load_dataset("tips")

# 1. Scatter Plot
plt.figure(figsize=(6,4))
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex", style="time")
plt.title("Scatter Plot - Total Bill vs Tip")
plt.show()

# 2. Line Plot
plt.figure(figsize=(6,4))
sns.lineplot(x="size", y="total_bill", data=tips, ci=None)
plt.title("Line Plot - Total Bill vs Size")
plt.show()

# 3. Bar Plot
plt.figure(figsize=(6,4))
sns.barplot(x="day", y="total_bill", data=tips, hue="sex")
plt.title("Bar Plot - Average Total Bill per Day")
plt.show()

# 4. Count Plot
plt.figure(figsize=(6,4))
sns.countplot(x="day", data=tips, hue="sex")
plt.title("Count Plot - Customers per Day")
plt.show()

# 5. Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex")
plt.title("Box Plot - Total Bill Distribution")
plt.show()

# 6. Violin Plot
plt.figure(figsize=(6,4))
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", split=True)
plt.title("Violin Plot - Distribution by Day and Gender")
plt.show()

# 7. Heatmap (Correlation Matrix)
plt.figure(figsize=(6,4))
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

# 8. Pairplot
sns.pairplot(tips, hue="sex")
plt.suptitle("Pairplot of Tips Dataset",y=1.02)
plt.show()

# 9. Distribution Plot (Histogram + KDE)
plt.figure(figsize=(6,4))
sns.histplot(tips["total_bill"], kde=True, bins=20)
plt.title("Distribution of Total Bill")
plt.show()

# 10. Regression Plot
plt.figure(figsize=(6,4))
sns.regplot(x="total_bill", y="tip", data=tips)
plt.title("Regression Plot - Tip vs Total Bill")
plt.show()
```

```python
# ==========================================
# Seaborn Implementation - Iris Dataset
# ==========================================

import seaborn as sns
import matplotlib.pyplot as plt

# Load Dataset
iris = sns.load_dataset("iris")

# ------------------------------------------
# Dataset Information
# ------------------------------------------

print("First 5 Rows:")
print(iris.head())

print("\nDataset Information:")
print(iris.info())

print("\nStatistical Summary:")
print(iris.describe())

print("\nShape of Dataset:", iris.shape)

print("\nColumn Names:")
print(iris.columns)

# ------------------------------------------
# 1. Scatter Plot
# Petal Length vs Petal Width
# ------------------------------------------

plt.figure(figsize=(6,4))

sns.scatterplot(
    x="petal_length",
    y="petal_width",
    data=iris,
    hue="species"
)

plt.title("Scatter Plot - Petal Length vs Petal Width")
plt.show()

# ------------------------------------------
# 2. Line Plot
# ------------------------------------------

plt.figure(figsize=(6,4))

sns.lineplot(
    x="sepal_length",
    y="petal_length",
    data=iris
)

plt.title("Line Plot - Sepal Length vs Petal Length")
plt.show()

# ------------------------------------------
# 3. Bar Plot
# Average Petal Length of Each Species
# ------------------------------------------

plt.figure(figsize=(6,4))

sns.barplot(
    x="species",
    y="petal_length",
    data=iris,
    hue="species"
)

plt.title("Bar Plot - Average Petal Length")
plt.show()

# ------------------------------------------
# 4. Count Plot
# Number of Flowers in Each Species
# ------------------------------------------

plt.figure(figsize=(6,4))

sns.countplot(
    x="species",
    data=iris,
    hue="species"
)

plt.title("Count Plot - Number of Flowers")
plt.show()

# ------------------------------------------
# 5. Box Plot
# ------------------------------------------

plt.figure(figsize=(6,4))

sns.boxplot(
    x="species",
    y="petal_width",
    data=iris,
    hue="species"
)

plt.title("Box Plot - Petal Width")
plt.show()

# ------------------------------------------
# 6. Violin Plot
# ------------------------------------------

plt.figure(figsize=(6,4))

sns.violinplot(
    x="species",
    y="petal_length",
    data=iris,
    hue="species"
)

plt.title("Violin Plot - Petal Length")
plt.show()

# ------------------------------------------
# 7. Heatmap
# Correlation Matrix
# ------------------------------------------

plt.figure(figsize=(6,4))

corr = iris.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Heatmap")
plt.show()

# ------------------------------------------
# 8. Pair Plot
# ------------------------------------------

sns.pairplot(
    iris,
    hue="species"
)

plt.suptitle("Pair Plot of Iris Dataset", y=1.02)

plt.show()

# ------------------------------------------
# 9. Histogram
# Distribution of Sepal Length
# ------------------------------------------

plt.figure(figsize=(6,4))

sns.histplot(
    iris["sepal_length"],
    kde=True,
    bins=15
)

plt.title("Distribution of Sepal Length")
plt.show()

# ------------------------------------------
# 10. Regression Plot
# ------------------------------------------

plt.figure(figsize=(6,4))

sns.regplot(
    x="petal_length",
    y="petal_width",
    data=iris
)

plt.title("Regression Plot - Petal Length vs Petal Width")
plt.show()

# ------------------------------------------
# 11. Cat Plot
# ------------------------------------------

sns.catplot(
    x="species",
    y="sepal_width",
    data=iris,
    kind="bar",
    hue="species"
)

plt.suptitle("Cat Plot - Average Sepal Width", y=1.02)

plt.show()

# ------------------------------------------
# Observation
# ------------------------------------------

print("\nObservation:")
print("Setosa is the most distinct species.")
print("It has much smaller petal length and petal width")
print("compared to Versicolor and Virginica.")
print("Versicolor and Virginica overlap slightly in most plots.")
```
