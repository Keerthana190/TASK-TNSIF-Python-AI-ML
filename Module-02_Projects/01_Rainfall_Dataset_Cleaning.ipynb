```python
print(df.head())

print(df.info())

print(df.describe())

print(df.isnull().sum())

print("Duplicate Records:", df.duplicated().sum())
```

Part 4: Convert Date Column

```python
df["date"] = df["date"].astype(str).str.strip()

df["date"] = pd.to_datetime(df["date"], errors="coerce")

df.info()
```

Part 5: Handle Missing Values

```python
num_cols = ["rainfall", "temperature", "humidity", "wind_speed"]

df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

df["weather_condition"] = df["weather_condition"].fillna(df["weather_condition"].mode()[0])

print(df.isnull().sum())
```

Part 6: Fix Text Inconsistencies

```python
df["weather_condition"] = df["weather_condition"].str.lower().str.strip()

print(df["weather_condition"].unique())
```

Part 7: Remove Duplicates

```python
print("Before:", df.duplicated().sum())

df = df.drop_duplicates()

print("After:", df.duplicated().sum())
```

Part 8: Detect Outliers (IQR)

```python
Q1 = df["rainfall"].quantile(0.25)
Q3 = df["rainfall"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["rainfall"] < lower) | (df["rainfall"] > upper)]

print(outliers)
```

Part 9: Handle Outliers

```python
median = df["rainfall"].median()

df["rainfall"] = np.where(df["rainfall"] < lower, median, df["rainfall"])
df["rainfall"] = np.where(df["rainfall"] > upper, median, df["rainfall"])

df.head()
```

Part 10: Validate Dataset

```python
print(df.info())

print(df.isnull().sum())

print(df.duplicated().sum())
```

Part 11: Save Dataset

```python
df.to_csv("cleaned_rainfall.csv", index=False)

print("Dataset Saved Successfully")
```
