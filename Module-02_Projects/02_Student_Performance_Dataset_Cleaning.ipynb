```python
import pandas as pd
import numpy as np

data = {
    "name": ["Amit", "Sara", "John", "Amit", None],
    "age": [20, 21, np.nan, 20, 19],
    "gender": ["Male", "female", "Male", "Male", "FEMALE"],
    "marks": [85, 90, 105, 85, -5],
    "attendance": [90, 85, 80, 90, None],
    "city": ["Mumbai", "Delhi", "mumbai", "Mumbai", "Pune"]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
df.isnull().sum()
df.describe()
```

## 🔹 Handling Missing Values

- age → mean
- attendance → mean
- name → "Unknown"

```python
df["age"].fillna(df["age"].mean(), inplace=True)
df["attendance"].fillna(df["attendance"].mean(), inplace=True)
df["name"].fillna("Unknown", inplace=True)

df
```

```python
df["gender"] = df["gender"].str.lower()
df["city"] = df["city"].str.lower()

df
```

## 🔹 Fix Invalid Data

Marks should be:
- Between 0 and 100

```python
df = df[(df["marks"] >= 0) & (df["marks"] <= 100)]

df
```

```python
print(df.head())

print(df.info())

print(df.describe())

print(df.isnull().sum())

print("Duplicate Records:", df.duplicated().sum())
```

Part 4: Handle Missing Values

```python
df["age"] = df["age"].fillna(df["age"].mean())

df["marks"] = df["marks"].fillna(df["marks"].mean())

df["attendance"] = df["attendance"].fillna(df["attendance"].mean())

df["gender"] = df["gender"].fillna(df["gender"].mode()[0])

df["city"] = df["city"].fillna(df["city"].mode()[0])

print(df.isnull().sum())
```

Part 5: Fix Text Inconsistencies

```python
df["gender"] = df["gender"].str.lower().str.strip()

df["city"] = df["city"].str.title().str.strip()

print(df["gender"].unique())

print(df["city"].unique())
```

Part 6: Remove Invalid Marks

```python
df = df[(df["marks"] >= 0) & (df["marks"] <= 100)]

df.head()
```

Part 7: Remove Duplicate Records

```python
print("Before:", df.duplicated().sum())

df = df.drop_duplicates()

print("After:", df.duplicated().sum())
```

Part 8: Validate Dataset

```python
print(df.info())

print(df.isnull().sum())

print(df.duplicated().sum())

print(df.describe())
```

Part 9 : Save Cleaned Dataset





```python
df.to_csv("cleaned_student_performance.csv", index=False)

print("Dataset Saved Successfully")
```
