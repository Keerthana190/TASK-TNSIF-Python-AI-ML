Import Libraries

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
```


Create Dataset

```python
data = {
    "Age": [22, 35, np.nan, 45, 52, 28, 31, np.nan, 40, 60],
    "Salary": [30000, 50000, 45000, np.nan, 120000, 38000, 42000, 46000, 55000, 500000],
    "Gender": ["Male", "Female", "Female", "Male", None, "Female", "Male", "Female", "Male", "Male"],
    "City": ["Hyderabad", "Delhi", "Mumbai", "Pune", "Delhi", "Hyderabad", None, "Mumbai", "Pune", "Delhi"],
    "Date": [
        "2023-01-15",
        "2022-05-10",
        "2021-08-25",
        "2023-03-18",
        "2020-12-05",
        "2022-11-20",
        "2021-07-09",
        "2023-06-30",
        "2022-09-14",
        "2020-01-01"
    ],
    "Purchase_Amount": [5000, 7000, 6500, 8000, 15000, 6000, 7500, 6800, 7200, 25000],
    "Experience": [1, 5, 4, 8, 15, 3, 6, 5, 7, 25],
    "Rating": [4.5, 4.2, 3.8, 4.7, 4.9, 3.9, 4.1, 4.0, 4.3, 5.0]
}

df = pd.DataFrame(data)

df
```

🔹 Problem 1 : Handling Missing Values

```python
print("Missing Values Before Filling")
print(df.isnull().sum())
```

```python
# Numerical columns

num_cols = ["Age","Salary"]

df[num_cols] = df[num_cols].fillna(df[num_cols].mean())
```

```python
# Categorical columns

cat_cols = ["Gender","City"]

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
print("Missing Values After Filling")
print(df.isnull().sum())

df
```

🔹 Problem 2 : Feature Scaling

```python
print("Before Standardization")
print(df["Salary"])

```

```python
scaler = StandardScaler()

df["Salary_Standardized"] = scaler.fit_transform(df[["Salary"]])

df[["Salary","Salary_Standardized"]]
```

```python
print("\nAfter Standardization")
print(df["Salary_Standardized"])
```

```python
print("Before Min-Max Scaling")
print(df["Age"])
```

```python
minmax = MinMaxScaler()

df["Age_Normalized"] = minmax.fit_transform(df[["Age"]])

df[["Age","Age_Normalized"]]
```

```python
print("\nAfter Min-Max Scaling")
print(df["Age_Normalized"])
```

🔹 Problem 3 : Encoding

```python
#Label Encoding

le = LabelEncoder()

df["Gender_Label"] = le.fit_transform(df["Gender"])

df
```

```python
#One-Hot Encoding

df_encoded = pd.get_dummies(df,columns=["City"])

df_encoded
```

🔹 Problem 4 : Feature Creation

```python
df["Total_Spending"] = df["Salary"] + df["Purchase_Amount"]

df
```

Analysis:

- Total_Spending combines Salary and Purchase_Amount.
- It gives the overall financial spending capability of a person.
- Instead of using two separate features, the model can also learn from this combined feature.
- This may improve prediction accuracy in some machine learning problems.

🔹 Problem 5 : Date Feature Extraction

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day
df["Weekday"] = df["Date"].dt.day_name()

df
```

🔹 Problem 6 : Binning

```python
bins = [0,25,45,100]

labels = ["Young","Adult","Senior"]

df["Age_Group"] = pd.cut(df["Age"],
                         bins=bins,
                         labels=labels)

df[["Age","Age_Group"]]
```

🔹 Problem 7 : Outlier Detection using IQR

```python
Q1 = df["Salary"].quantile(0.25)

Q3 = df["Salary"].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - (1.5 * IQR)

upper_limit = Q3 + (1.5 * IQR)
```

```python
outliers = df[(df["Salary"] < lower_limit) |
              (df["Salary"] > upper_limit)]

print("Outlier Rows")

outliers
```

```python
outliers

print("Number of Outliers:", len(outliers))
```

🔹 Problem 8 : Feature Interaction

```python
df["Experience_Salary"] = df["Experience"] * df["Salary"]

df
```

Importance of Feature Interaction

- Experience_Salary combines two important features.
- It captures the relationship between salary and work experience.
- It helps the model understand whether experienced employees generally earn higher salaries.
- Such interaction features can improve prediction accuracy because they provide additional information beyond individual features.

🔹 Problem 9 : Feature Selection

```python
numeric_df = df.select_dtypes(include=np.number)

numeric_df
```

```python
#Correlation Matrix
corr_matrix = numeric_df.corr()

corr_matrix
```

```python
#Correlation with Purchase Amount
corr_purchase = corr_matrix["Purchase_Amount"]

print(corr_purchase)
```

```python
print("Correlation with Purchase_Amount")
print(corr_purchase.sort_values(ascending=False))
```

```python
#Drop Least Important Feature
least_feature = corr_purchase.drop("Purchase_Amount").abs().idxmin()

print("Least Important Feature :",least_feature)
```

```python
df_selected = df.drop(columns=[least_feature])

df_selected
```

```python
print("Final Dataset after Feature Selection")
df_selected
```
